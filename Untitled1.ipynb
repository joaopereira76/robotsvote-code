```python
from reachy_mini import ReachyMini
import soundfile as sf
import time

with ReachyMini() as mini:
    # 1. Carregar o ficheiro (precisa da biblioteca soundfile instalada)
    # O Reachy Mini trabalha melhor com ficheiros .wav
    data, samplerate = sf.read('test1.wav') 

    # 2. Iniciar o sistema de som
    mini.media.start_playing()

    # 3. Enviar o áudio para o robô 48000Hz
    mini.media.push_audio_sample(data)

    # 4. Esperar o áudio terminar antes de fechar o código
    duracao = len(data) / samplerate
    time.sleep(duracao)

    # 5. Parar o sistema de som
    mini.media.stop_playing()
```

```python
from reachy_mini import ReachyMini
import soundfile as sf
import time
import random
from reachy_mini.utils import create_head_pose

# Dicionário para guardar as instâncias
robos = {}

def conectar_robos():
    # Inicializa sem o 'with' para a conexão não fechar
    r = ReachyMini()
    # Inicia o sistema de áudio apenas UMA vez
    r.media.start_playing()
    return r

# Conecta uma vez e guarda
robos["Francisca"] = conectar_robos()

def animar_fala(robot, duracao):
    """Move a cabeça sutilmente enquanto fala (Versão Síncrona)."""
    # Define o tempo de término
    fim = time.time() + duracao
    
    while time.time() < fim:
        # Pequenos movimentos aleatórios
        r = random.uniform(-20, 20)  # Roll
        print("Passou aqui " + str(r))
        
        
        robot.goto_target(
            head=create_head_pose(z=0, roll=r, degrees=True),
            duration=2
        )
        
    
    
def tocar_audio_imediato(robo, caminho_ficheiro):
    data, samplerate = sf.read(caminho_ficheiro)
    # Envia diretamente para o buffer (sem abrir/fechar o sistema de som)
    robo.media.push_audio_sample(data)
   
    animar_fala(robo, len(data)/48000)
    



# --- Agora podes chamar isto quando quiseres, sem delay ---
tocar_audio_imediato(robos["Francisca"], "test1.wav")
```
