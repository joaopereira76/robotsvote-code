```python
import os
import wave
import time
import random
import numpy as np
import soundfile as sf
from reachy_mini import ReachyMini
from reachy_mini.utils import create_head_pose

# --- Parâmetros ---
NETWORK_LEAD   = 0.10   # Latência de rede estimada (segundos) — ajustar com ping
MAX_PITCH      = 12     # Graus máx. para cima/baixo (mais subtil)
MAX_ROLL       = 10     # Graus máx. de inclinação lateral
MAX_YAW        = 14     # Graus máx. de rotação horizontal
RMS_WINDOW     = 0.4    # Janela para calcular volume médio (segundos)

# Duração das poses em função do volume (segundos)
POSE_DUR_QUIET = (1.5, 3.0)   # silêncio ou fala suave
POSE_DUR_MID   = (0.9, 1.6)   # fala normal
POSE_DUR_LOUD  = (0.5, 0.9)   # fala alta ou ênfase


def _rms_envelope(data: np.ndarray, samplerate: int, window: float) -> np.ndarray:
    """RMS por janela — devolve um valor por janela ao longo do áudio."""
    n = int(samplerate * window)
    frames = len(data) // n
    env = np.zeros(frames)
    for i in range(frames):
        chunk = data[i * n : (i + 1) * n]
        env[i] = float(np.sqrt(np.mean(chunk ** 2)))
    return env


def _random_pose(prev: tuple, intensity: float) -> tuple:
    """Gera uma nova pose suave, nunca oposta à anterior.

    Usa um offset em relação à pose anterior em vez de valores completamente
    aleatórios, para garantir continuidade e evitar saltos bruscos.
    """
    scale = 0.3 + 0.7 * min(intensity * 4, 1.0)  # 0.3 em silêncio, 1.0 no máximo

    def step(prev_val, max_angle):
        # Deslocamento máximo proporcional à escala e ao ângulo máximo
        delta = random.uniform(-max_angle * 0.5, max_angle * 0.5) * scale
        new_val = prev_val + delta
        # Mantém dentro dos limites
        return float(np.clip(new_val, -max_angle, max_angle))

    prev_pitch, prev_roll, prev_yaw = prev
    return (
        step(prev_pitch, MAX_PITCH),
        step(prev_roll,  MAX_ROLL),
        step(prev_yaw,   MAX_YAW),
    )


def _build_keyframes(data: np.ndarray, samplerate: int, wav_duration: float):
    """Constrói a lista de keyframes (timestamp, pose) antes de tocar.

    A duração de cada pose depende do volume nesse momento:
    - silêncio: pose dura mais (robô quase parado)
    - fala alta: pose dura menos (robô mais expressivo)
    """
    env = _rms_envelope(data, samplerate, RMS_WINDOW)
    # Normaliza o envelope entre 0 e 1
    env_max = env.max() if env.max() > 0 else 1.0
    env_norm = env / env_max

    keyframes = []  # lista de (t_início, pitch, roll, yaw, duração)
    t = 0.0
    prev_pose = (0.0, 0.0, 0.0)  # começa centrado

    while t < wav_duration:
        # Volume aproximado neste momento
        env_idx = min(int(t / RMS_WINDOW), len(env_norm) - 1)
        intensity = float(env_norm[env_idx])

        # Duração desta pose em função do volume
        if intensity < 0.25:
            dur = random.uniform(*POSE_DUR_QUIET)
        elif intensity < 0.65:
            dur = random.uniform(*POSE_DUR_MID)
        else:
            dur = random.uniform(*POSE_DUR_LOUD)

        # Não ultrapassar o fim do áudio
        dur = min(dur, wav_duration - t)

        pose = _random_pose(prev_pose, intensity)
        keyframes.append((t, pose, dur))
        prev_pose = pose
        t += dur

    return keyframes


def play_audio_keyframe(mini, wav_path: str):
    wav_path = os.path.abspath(wav_path)

    # 1. Carrega e analisa antes de tocar
    data, samplerate = sf.read(wav_path)
    if data.ndim > 1:
        data = data.mean(axis=1)

    with wave.open(wav_path, "rb") as wf:
        wav_duration = wf.getnframes() / mini.media.get_output_audio_samplerate()

    keyframes = _build_keyframes(data, samplerate, wav_duration)
    print(f"A tocar '{os.path.basename(wav_path)}' ({wav_duration:.1f}s, {len(keyframes)} poses)")

    # 2. Começa a tocar o áudio
    mini.media.play_sound(wav_path)
    t0 = time.monotonic()

    for (kf_t, pose, dur) in keyframes:
        pitch, roll, yaw = pose

        # Envia o goto_target com NETWORK_LEAD de antecedência
        send_at = t0 + kf_t - NETWORK_LEAD
        wait = send_at - time.monotonic()
        if wait > 0:
            time.sleep(wait)

        # Verifica se o áudio já acabou
        if time.monotonic() - t0 >= wav_duration:
            break

        mini.goto_target(
            head=create_head_pose(
                pitch=pitch,
                roll=roll,
                yaw=yaw,
                degrees=True,
            ),
            duration=dur,  # O robô interpola durante exatamente esta duração
        )

    # Reset suave
    mini.cancel_move()
    mini.goto_target(head=create_head_pose(), duration=1.5)
    print("Fim.")
    time.sleep(1.5)


# --- Uso ---
with ReachyMini(host="reachy-mini-one.local") as mini:
    play_audio_keyframe(mini, "test1.wav")
```

```python
import time
import numpy as np
import soundfile as sf
from reachy_mini import ReachyMini
from reachy_mini.utils import create_head_pose

AUDIO_FILE = "test1.wav"

STEP_TIME = 0.2   # 🔥 MUITO importante (5 Hz)
SENSITIVITY = 1.0


def play_audio_with_head_motion(robot, file_path):
    data, samplerate = sf.read(file_path)

    if len(data.shape) > 1:
        data = data.mean(axis=1)

    robot.media.start_playing()

    samples_per_step = int(samplerate * STEP_TIME)

    for i in range(0, len(data), samples_per_step):
        start_time = time.monotonic()

        chunk = data[i:i + samples_per_step]

        # 🎧 enviar áudio
        robot.media.push_audio_sample(chunk)

        if len(chunk) > 0:
            rms = np.sqrt(np.mean(chunk**2))

            pitch = (rms ** 0.5) * SENSITIVITY * 50
            print(pitch)
            roll = np.random.uniform(-50, 50) * rms
            yaw = np.random.uniform(-50, 50) * rms

            # ✅ movimento LENTO (sem spam!)
            robot.goto_target(
                head=create_head_pose(
                    roll=roll,
                    pitch=pitch,
                    yaw=yaw,
                    degrees=True
                ),
                duration=STEP_TIME  # igual ao intervalo!
            )

        # ⏱️ sync
        elapsed = time.monotonic() - start_time
        sleep_time = STEP_TIME - elapsed

        if sleep_time > 0:
            time.sleep(sleep_time)

    # reset
    robot.goto_target(head=create_head_pose(), duration=1.0)


if __name__ == "__main__":
    with ReachyMini(host="reachy-mini-one.local") as mini:
        print("Ligado")

        mini.goto_target(head=create_head_pose(), duration=1.0)

        play_audio_with_head_motion(mini, AUDIO_FILE)

        print("Done")
```

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# =========================
# CONFIG
# =========================



# =========================
# CENTER CROP (IMPORTANTE)
# =========================
def crop_center(img, target_w, target_h):
    img_ratio = img.width / img.height
    target_ratio = target_w / target_h

    # cortar largura
    if img_ratio > target_ratio:
        new_width = int(img.height * target_ratio)
        offset = (img.width - new_width) // 2
        img = img.crop((offset, 0, offset + new_width, img.height))
    else:
        new_height = int(img.width / target_ratio)
        offset = (img.height - new_height) // 2
        img = img.crop((0, offset, img.width, offset + new_height))

    return img.resize((target_w, target_h))


# =========================
# WRAP TEXTO
# =========================
def wrap_text(text, font, max_width, draw):
    lines = []
    words = text.split()

    while words:
        line = ""
        while words:
            test_line = line + words[0] + " "
            bbox = draw.textbbox((0, 0), test_line, font=font)
            w = bbox[2] - bbox[0]

            if w <= max_width:
                line = test_line
                words.pop(0)
            else:
                break

        lines.append(line.strip())

    return lines


def criar_card(titulo, quote, imagem_path, output):

    WIDTH, HEIGHT = 540, 540
    
    img = Image.open("./articles/background/economic.jpg").convert("RGB")
    img = crop_center(img, WIDTH, HEIGHT)

    # escurecer ligeiramente (melhor leitura)
    black = Image.new("RGB", (WIDTH, HEIGHT), (0, 0, 0))
    img = Image.blend(img, black, 0.8)

    draw = ImageDraw.Draw(img)

    titulo_font = ImageFont.truetype("./Lato/Lato-Bold.ttf", 30)
    quote_font = ImageFont.truetype("./Lato/Lato-Regular.ttf", 20)

    max_width = 500

    titulo_lines = wrap_text(titulo, titulo_font, max_width, draw)

    y = 50

    for line in titulo_lines:
        bbox = draw.textbbox((0, 0), line, font=titulo_font)
        text_width = bbox[2] - bbox[0]

        x = (WIDTH - text_width) / 2

        draw.text((x, y), line, font=titulo_font, fill="white")
        y += 50

    quote_text = f"“{quote}”"
    quote_lines = wrap_text(quote_text, quote_font, max_width, draw)

    line_height = 25
    total_height = len(quote_lines) * line_height

    y_quote = HEIGHT - total_height - 90

    for line in quote_lines:
        bbox = draw.textbbox((0, 0), line, font=quote_font)
        text_width = bbox[2] - bbox[0]

        x = (WIDTH - text_width) / 2

        draw.text((x, y_quote), line, font=quote_font, fill="white")
        y_quote += line_height

    img.save(output)
    print(f"✔ Guardado: {output}")

# Dados para gerar 4 imagens
cards = [
    ("Robôs como Facilitadores de Ligação", 'Discordo totalmente, em vez de isolarem, os robôs podem atuar como facilitadores, ajudando pessoas que já estão isoladas a ligarem-se melhor ao mundo e a outros humanos. A tecnologia pode servir de ponte para quem tem dificuldades sociais, funcionando como um treino para interações reais. O robô não substitui o humano, mas prepara o terreno para novas socializações.', "#1a1a1a"),
    ("Breaking News", "Algo importante aconteceu", "#333366"),
    ("Atualidade", "Frase impactante aqui", "#660000"),
    ("Tecnologia", "O futuro chegou", "#003366"),
]

# Gerar imagens
for i, (titulo, quote, cor) in enumerate(cards):
    criar_card(titulo, quote, cor, f"./articles/card_{i+1}.png")
```
