```python
import asyncio
import json
import threading
import uuid
import webbrowser
from datetime import datetime
from http.server import BaseHTTPRequestHandler, HTTPServer
from pathlib import Path

RESPONSES_FILE = "responses.json"

# ── Dados dos questionários ───────────────────────────────────────────────────

TOPICS = {
    "economic": {
        "label": "Posições Socioeconómicas — Económico",
        "options": ["Discordo totalmente", "Discordo", "Discordo ligeiramente",
                    "Concordo ligeiramente", "Concordo", "Concordo totalmente"],
        "statements": [
            "O mérito pessoal é suficiente para se ter sucesso em Portugal.",
            "O politicamente correto tornou o mundo monótono.",
            "A qualidade dos serviços de saúde e da educação deve ser garantida pelo governo.",
            "O setor privado é um gestor mais honesto do que o governo.",
            "Quanto menos o governo interferir na sociedade, melhor será o país.",
            "A igualdade social é mais importante do que a liberdade económica.",
            "Os criminosos devem ser punidos com a pena de morte.",
            "O governo deve regulamentar o setor privado para promover o interesse público.",
        ]
    },
    "social": {
        "label": "Posições Socioeconómicas — Social",
        "options": ["Discordo totalmente", "Discordo", "Discordo ligeiramente",
                    "Concordo ligeiramente", "Concordo", "Concordo totalmente"],
        "statements": [
            "As mulheres devem ter o direito de optar pelo aborto.",
            "A legalização e a regulamentação das drogas leves trariam resultados positivos.",
            "Os movimentos em defesa dos direitos das minorias são legítimos.",
            "Apenas os casais heterossexuais devem poder adotar.",
            "A separação entre a Igreja e o Estado em questões jurídicas e políticas é essencial.",
            "O aumento do número de imigrantes em Portugal é necessário e positivo.",
        ]
    },
    "robot_related": {
        "label": "Atitudes em relação aos Robôs",
        "options": ["Discordo totalmente", "Discordo", "Discordo ligeiramente",
                    "Concordo ligeiramente", "Concordo", "Concordo totalmente"],
        "statements": [
            "Os robôs são necessários porque conseguem realizar tarefas demasiado difíceis ou perigosas para as pessoas.",
            "Os robôs podem facilitar a nossa vida.",
            "Atribuir tarefas de rotina aos robôs permite que as pessoas se dediquem a tarefas mais significativas.",
            "As tarefas perigosas devem ser atribuídas, em primeiro lugar, aos robôs.",
            "Os robôs são benéficos para a sociedade, porque ajudam as pessoas.",
            "Os robôs podem tornar-nos ainda mais preguiçosos.",
            "A utilização generalizada de robôs vai tirar empregos às pessoas.",
            "Receio que os robôs incentivem uma menor interação entre os seres humanos.",
            "A robótica é uma das áreas da tecnologia que precisa de ser rigorosamente supervisionada.",
            "A utilização não regulamentada da robótica pode levar a mudanças sociais.",
        ]
    }
}

KNOWLEDGE_BLOCK = {
    "label": "Conhecimento sobre esta afirmação",
    "options": ["Não, de todo", "Muito pouco", "Pouco", "Moderadamente",
                "Neutro", "Razoavelmente", "Bastante", "Muito", "Extremamente"],
    "statements": [
        "Em que medida considera-se bem informado sobre este tema?",
        "Em que medida considera que possui informação sobre este tema?",
        "Em que medida considera que possui conhecimentos especializados sobre este tema?",
        "Em que medida considera que conhece os factos mais relevantes sobre este tema?",
    ]
}

DEMOGRAPHICS_PRE = [
    {"id": "age",         "label": "Idade",               "type": "number"},
    {"id": "gender",      "label": "Identidade de género", "type": "text"},
    {"id": "nationality", "label": "Nacionalidade",        "type": "text"},
    {"id": "university",  "label": "Universidade",         "type": "text"},
    {"id": "course",      "label": "Área de estudo/Curso", "type": "text"},
]

DEMOGRAPHICS_POS = [
    {"id": "party",      "label": "Em que partido votou nas últimas eleições legislativas?", "type": "text"},
    {"id": "ai_freq",    "label": "Nos últimos 6 meses, com que frequência utilizou a IA?",
     "type": "select", "options": ["Não utilizei nos últimos seis meses", "Menos de uma vez por mês",
                                    "Cerca de uma vez por mês", "Cerca de duas vezes por mês",
                                    "Cerca de três vezes por mês", "Semanalmente", "Diariamente"]},
    {"id": "robot_freq", "label": "Nos últimos 6 meses, com que frequência interagiu com robôs?",
     "type": "select", "options": ["Não interagi nos últimos seis meses", "Menos de uma vez por mês",
                                    "Cerca de uma vez por mês", "Cerca de duas vezes por mês",
                                    "Cerca de três vezes por mês", "Semanalmente", "Diariamente"]},
    {"id": "robot_types", "label": "Com que tipos de robôs já interagiste?", "type": "text"},
]


# ── Carregar / guardar respostas ──────────────────────────────────────────────

def load_responses():
    if Path(RESPONSES_FILE).exists():
        with open(RESPONSES_FILE, "r", encoding="utf-8") as f:
            return json.load(f)
    return {}

def save_responses(data):
    with open(RESPONSES_FILE, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)


# ── Gerar HTML ────────────────────────────────────────────────────────────────

def build_html(phase, topic, participant_id):
    is_pre = phase == "pre"
    title = "Questionário Inicial" if is_pre else "Questionário Final"
    topic_data = TOPICS[topic]

    def likert_block_with_knowledge(block_label, statements, stmt_options, topic):
        """
        Renders each statement followed immediately by the 4 knowledge sub-questions.
        Each statement+knowledge unit is visually grouped inside a card.
        """
        k_opts = KNOWLEDGE_BLOCK["options"]
        k_stmts = KNOWLEDGE_BLOCK["statements"]

        # Header row for the main statement options
        stmt_opt_headers = "".join(
            f'<th class="opt-col">{o}</th>' for o in stmt_options
        )
        # Header row for the knowledge options
        k_opt_headers = "".join(
            f'<th class="opt-col k-opt">{o}</th>' for o in k_opts
        )

        cards = ""
        for i, stmt in enumerate(statements):
            stmt_idx = i + 1

            # --- Main statement row ---
            stmt_cells = "".join(
                f'<td><label class="radio-wrap"><input type="radio" name="q_{topic}_{stmt_idx}" value="{j+1}" required><span class="dot"></span></label></td>'
                for j in range(len(stmt_options))
            )

            # --- Knowledge sub-rows ---
            k_rows = ""
            for ki, k_stmt in enumerate(k_stmts):
                k_cells = "".join(
                    f'<td><label class="radio-wrap"><input type="radio" name="q_knowledge_{topic}_{stmt_idx}_{ki+1}" value="{kj+1}" required><span class="dot dot-k"></span></label></td>'
                    for kj in range(len(k_opts))
                )
                k_rows += f"""
                <tr class="k-row">
                  <td class="k-stmt-text">{k_stmt}</td>
                  {k_cells}
                </tr>"""

            cards += f"""
            <div class="stmt-card">
              <div class="stmt-card-header">
                <span class="stmt-num">{stmt_idx}</span>
                <span class="stmt-card-text">{stmt}</span>
              </div>

              <div class="table-wrap">
                <table class="likert">
                  <thead>
                    <tr>
                      <th class="stmt-text-col"></th>
                      {stmt_opt_headers}
                    </tr>
                  </thead>
                  <tbody>
                    <tr class="stmt-row main-row">
                      <td class="stmt-text">Grau de concordância</td>
                      {stmt_cells}
                    </tr>
                  </tbody>
                </table>
              </div>

              <div class="k-divider">
                <span class="k-label">{KNOWLEDGE_BLOCK["label"]}</span>
              </div>

              <div class="table-wrap">
                <table class="likert k-table">
                  <thead>
                    <tr>
                      <th class="stmt-text-col"></th>
                      {k_opt_headers}
                    </tr>
                  </thead>
                  <tbody>{k_rows}</tbody>
                </table>
              </div>
            </div>"""

        return f"""
        <div class="block">
          <h2 class="block-title">{block_label}</h2>
          {cards}
        </div>"""

    # Bloco de demographics
    def demo_block(fields):
        inputs = ""
        for f in fields:
            if f["type"] == "select":
                opts = "".join(f'<option value="{o}">{o}</option>' for o in f["options"])
                inputs += f'<div class="field"><label>{f["label"]}</label><select name="{f["id"]}" required><option value="">-- seleciona --</option>{opts}</select></div>'
            else:
                inputs += f'<div class="field"><label>{f["label"]}</label><input type="{f["type"]}" name="{f["id"]}" required></div>'
        return f'<div class="block"><h2 class="block-title">Dados Pessoais</h2>{inputs}</div>'

    # Montar blocos consoante fase
    if is_pre:
        blocks = (
            demo_block(DEMOGRAPHICS_PRE) +
            likert_block_with_knowledge(topic_data["label"], topic_data["statements"], topic_data["options"], topic)
        )
    else:
        blocks = (
            likert_block_with_knowledge(topic_data["label"], topic_data["statements"], topic_data["options"], topic) +
            demo_block(DEMOGRAPHICS_POS)
        )

    return f"""<!DOCTYPE html>
<html lang="pt">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=IBM+Plex+Mono:wght@400;600&family=IBM+Plex+Sans:wght@400;500;600&display=swap" rel="stylesheet">
<style>
*, *::before, *::after {{ box-sizing: border-box; margin: 0; padding: 0; }}

:root {{
  --bg: #f4f1ec;
  --surface: #ffffff;
  --border: #d6d0c8;
  --text: #1a1816;
  --muted: #7a746e;
  --accent: #2a5298;
  --accent-light: #e8eef7;
  --dot: #2a5298;
  --dot-k: #7a5c2a;
  --k-bg: #faf7f2;
  --k-border: #e0d8cc;
  --radius: 8px;
}}

body {{
  font-family: 'IBM Plex Sans', sans-serif;
  background: var(--bg);
  color: var(--text);
  min-height: 100vh;
  padding: 40px 20px 80px;
}}

.container {{
  max-width: 960px;
  margin: 0 auto;
}}

header {{
  margin-bottom: 40px;
  padding-bottom: 24px;
  border-bottom: 2px solid var(--text);
}}

.phase-tag {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 11px;
  font-weight: 600;
  letter-spacing: 0.15em;
  text-transform: uppercase;
  color: var(--muted);
  margin-bottom: 8px;
}}

h1 {{
  font-size: 28px;
  font-weight: 600;
  line-height: 1.2;
}}

.pid {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 11px;
  color: var(--muted);
  margin-top: 8px;
}}

/* Blocks */
.block {{
  background: var(--surface);
  border: 1px solid var(--border);
  border-radius: var(--radius);
  padding: 28px;
  margin-bottom: 24px;
}}

.block-title {{
  font-size: 13px;
  font-weight: 600;
  letter-spacing: 0.08em;
  text-transform: uppercase;
  color: var(--muted);
  margin-bottom: 20px;
  padding-bottom: 12px;
  border-bottom: 1px solid var(--border);
}}

/* Statement cards */
.stmt-card {{
  border: 1px solid var(--border);
  border-radius: var(--radius);
  margin-bottom: 20px;
  overflow: hidden;
}}

.stmt-card:last-child {{
  margin-bottom: 0;
}}

.stmt-card-header {{
  display: flex;
  align-items: flex-start;
  gap: 12px;
  padding: 16px 20px;
  background: var(--bg);
  border-bottom: 1px solid var(--border);
}}

.stmt-num {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 12px;
  font-weight: 600;
  color: var(--accent);
  min-width: 20px;
  padding-top: 2px;
}}

.stmt-card-text {{
  font-size: 15px;
  font-weight: 500;
  line-height: 1.5;
}}

/* Likert table */
.table-wrap {{ overflow-x: auto;}}

.likert {{
  width: 100%;
  border-collapse: collapse;
  font-size: 13px;
}}

.likert thead th {{
  font-size: 11px;
  font-weight: 500;
  color: var(--muted);
  text-align: center;
  padding: 10px 6px 8px;
  line-height: 1.3;
}}

.stmt-text-col {{ width: 38%; text-align: left !important; }}

.stmt-row {{ border-top: 1px solid var(--border); }}
.main-row td {{ padding: 12px 4px; }}

.stmt-text {{
  padding: 12px 12px 12px 0;
  line-height: 1.5;
  font-size: 13px;
  vertical-align: middle;
  color: var(--muted);
  font-style: italic;
}}

.likert td {{
  text-align: center;
  vertical-align: middle;
}}

/* Knowledge section */
.k-divider {{
  display: flex;
  align-items: center;
  gap: 10px;
  padding: 10px 20px;
  background: var(--k-bg);
  border-top: 1px solid var(--k-border);
  border-bottom: 1px solid var(--k-border);
}}

.k-label {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 10px;
  font-weight: 600;
  letter-spacing: 0.1em;
  text-transform: uppercase;
  color: var(--dot-k);
}}

.k-table {{
  background: var(--k-bg);
}}

.k-table thead th {{
  color: var(--dot-k);
  opacity: 0.8;
}}

.k-opt {{ min-width: 56px; }}

.k-row {{ border-top: 1px solid var(--k-border); }}
.k-row td {{ padding: 10px 4px; }}

.k-stmt-text {{
  padding: 10px 10px 10px 0;
  font-size: 12px;
  line-height: 1.4;
  color: var(--muted);
  vertical-align: middle;
}}

/* Radio dots */
.radio-wrap {{ cursor: pointer; display: inline-flex; }}
.radio-wrap input {{ display: none; }}

.dot {{
  width: 18px; height: 18px;
  border-radius: 50%;
  border: 2px solid var(--border);
  display: block;
  transition: all .15s;
  background: white;
}}

.radio-wrap input:checked + .dot {{
  background: var(--dot);
  border-color: var(--dot);
  box-shadow: 0 0 0 3px var(--accent-light);
}}

.radio-wrap:hover .dot {{ border-color: var(--accent); }}

.dot-k {{
  width: 16px; height: 16px;
  border-radius: 50%;
  border: 2px solid var(--k-border);
  background: white;
}}

.radio-wrap input:checked + .dot-k {{
  background: var(--dot-k);
  border-color: var(--dot-k);
  box-shadow: 0 0 0 3px #f0e8d8;
}}

.radio-wrap:hover .dot-k {{ border-color: var(--dot-k); }}

/* Demographics fields */
.field {{ margin-bottom: 18px; }}
.field:last-child {{ margin-bottom: 0; }}

.field label {{
  display: block;
  font-size: 13px;
  font-weight: 500;
  margin-bottom: 6px;
  color: var(--text);
}}

.field input, .field select {{
  width: 100%;
  padding: 10px 12px;
  border: 1px solid var(--border);
  border-radius: var(--radius);
  font-family: 'IBM Plex Sans', sans-serif;
  font-size: 14px;
  background: var(--bg);
  color: var(--text);
  transition: border-color .15s;
}}

.field input:focus, .field select:focus {{
  outline: none;
  border-color: var(--accent);
  background: white;
}}

/* Submit */
.submit-wrap {{
  text-align: center;
  margin-top: 32px;
}}

.btn {{
  font-family: 'IBM Plex Sans', sans-serif;
  font-size: 15px;
  font-weight: 600;
  padding: 14px 48px;
  background: var(--text);
  color: white;
  border: none;
  border-radius: var(--radius);
  cursor: pointer;
  transition: opacity .15s;
}}
.btn:hover {{ opacity: 0.85; }}
.btn:disabled {{ opacity: 0.4; cursor: not-allowed; }}

.msg-ok {{
  display: none;
  margin-top: 20px;
  font-size: 14px;
  color: #2a7a2a;
  font-weight: 500;
}}
</style>
</head>
<body>
<div class="container">
  <header>
    <div class="phase-tag">{"Questionário Inicial" if is_pre else "Questionário Final"} &mdash; {topic_data["label"]}</div>
    <h1>{"Bem-vindo ao estudo." if is_pre else "Estamos quase a terminar."}</h1>
    <div class="pid">ID: {participant_id}</div>
  </header>

  <form id="qform">
    <input type="hidden" name="participant_id" value="{participant_id}">
    <input type="hidden" name="phase" value="{phase}">
    <input type="hidden" name="topic" value="{topic}">
    {blocks}
    <div class="submit-wrap">
      <button type="submit" class="btn" id="submit-btn">Submeter respostas</button>
      <div class="msg-ok" id="msg-ok">Respostas guardadas. Pode fechar esta janela.</div>
    </div>
  </form>
</div>

<script>
document.getElementById('qform').addEventListener('submit', async function(e) {{
  e.preventDefault();
  const btn = document.getElementById('submit-btn');
  btn.disabled = true;
  btn.textContent = 'A guardar...';

  const fd = new FormData(this);
  const data = {{}};
  for (const [k, v] of fd.entries()) data[k] = v;

  // Recolher radios manualmente (FormData so inclui os checked)
  document.querySelectorAll('input[type=radio]:checked').forEach(r => {{
    data[r.name] = r.value;
  }});

  try {{
    const r = await fetch('/submit', {{
      method: 'POST',
      headers: {{'Content-Type': 'application/json'}},
      body: JSON.stringify(data)
    }});
    if (r.ok) {{
      document.getElementById('msg-ok').style.display = 'block';
      btn.textContent = 'Submetido';
    }} else {{
      btn.disabled = false;
      btn.textContent = 'Submeter respostas';
      alert('Erro ao guardar. Tenta novamente.');
    }}
  }} catch(err) {{
    btn.disabled = false;
    btn.textContent = 'Submeter respostas';
    alert('Erro de ligacao: ' + err.message);
  }}
}});
</script>
</body>
</html>"""


# ── Handler HTTP ──────────────────────────────────────────────────────────────

def make_handler_quest(phase, topic, participant_id, done_event):
    class Handler(BaseHTTPRequestHandler):
        def log_message(self, fmt, *args): pass

        def do_GET(self):
            html = build_html(phase, topic, participant_id).encode("utf-8")
            self.send_response(200)
            self.send_header("Content-Type", "text/html; charset=utf-8")
            self.end_headers()
            self.wfile.write(html)

        def do_POST(self):
            if self.path == "/submit":
                length = int(self.headers.get("Content-Length", 0))
                body = json.loads(self.rfile.read(length))

                # Guardar no JSON
                all_data = load_responses()
                pid = body.get("participant_id", participant_id)
                if phase == "pre":
                    if pid not in all_data:
                        all_data[pid] = {}
                    all_data[pid] = {
                        "timestamp": datetime.now().isoformat(),
                        "topic": topic,
                        "pre": body
                    }
                else:
                    all_data[pid]["pro"] = body
                save_responses(all_data)
                print(f"  -> Respostas do {phase} guardadas para {pid}")

                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.end_headers()
                self.wfile.write(b'{"ok": true}')

                # Sinalizar que terminou
                done_event.set()
            else:
                self.send_response(404)
                self.end_headers()

    return Handler


# ── Ponto de entrada ──────────────────────────────────────────────────────────

def run_questionnaire_sync(phase, topic, participant_id=None, port=8001):
    import uuid, threading, webbrowser
    from http.server import HTTPServer

    if participant_id is None:
        participant_id = str(uuid.uuid4())

    done_event = threading.Event()
    handler = make_handler_quest(phase, topic, participant_id, done_event)
    server = HTTPServer(("0.0.0.0", port), handler)

    t = threading.Thread(target=server.serve_forever, daemon=True)
    t.start()

    webbrowser.open(f"http://127.0.0.1:{port}")
    print(f"Abre http://127.0.0.1:{port} se o browser nao abrir sozinho")

    done_event.wait()  # bloqueia até o participante submeter
    server.shutdown()
    print(f"Questionario {phase} concluido.")
    return participant_id
```

```python
"""
gerar_capa_jornal.py
--------------------
Gera uma capa de jornal estilizada como imagem PNG.
- Manchete, lead, imagem e legenda são personalizáveis
- Cabeçalho do jornal e colunas laterais ficam com blur
  para simular um jornal real sem revelar a fonte

Dependências: Pillow
  pip install Pillow

Fontes usadas (Liberation Serif/Sans — já incluídas em
sistemas Linux; em Windows/Mac substitua pelos caminhos
das tuas fontes).

Uso:
  python gerar_capa_jornal.py
  ou importe e chame gerar_capa() com os teus parâmetros.
"""

from PIL import Image, ImageDraw, ImageFont, ImageFilter
import textwrap
import os
import datetime

# ── Caminhos de fontes (ajusta se necessário) ──────────────────────────────
SERIF_BOLD   = "/usr/share/fonts/truetype/liberation/LiberationSerif-Bold.ttf"
SERIF_REG    = "/usr/share/fonts/truetype/liberation/LiberationSerif-Regular.ttf"
SERIF_ITALIC = "/usr/share/fonts/truetype/liberation/LiberationSerif-Italic.ttf"
SANS_BOLD    = "/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf"
SANS_REG     = "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf"

# ── Paleta ─────────────────────────────────────────────────────────────────
PAPER   = (249, 247, 242)   # fundo bege-papel
INK     = (15,  15,  15)    # texto principal
MUTED   = (90,  85,  75)    # texto secundário
RULE    = (30,  30,  30)    # linhas de separação
PULLBG  = (238, 234, 216)   # fundo da caixa de destaque


# ── Utilitários ────────────────────────────────────────────────────────────

def load_font(path, size):
    try:
        return ImageFont.truetype(path, size)
    except OSError:
        return ImageFont.load_default()


def wrap_lines(draw, text, font, max_width):
    """Divide texto em linhas que cabem em max_width pixels."""
    words = text.split()
    lines, current = [], ""
    for word in words:
        test = (current + " " + word).strip()
        w = draw.textlength(test, font=font)
        if w <= max_width:
            current = test
        else:
            if current:
                lines.append(current)
            current = word
    if current:
        lines.append(current)
    return lines


def draw_text_block(draw, lines, font, x, y, fill, line_spacing=1.25):
    """Desenha um bloco de linhas, devolve y final."""
    bbox_h = draw.textbbox((0, 0), "Ag", font=font)
    line_h = int((bbox_h[3] - bbox_h[1]) * line_spacing)
    for line in lines:
        draw.text((x, y), line, font=font, fill=fill)
        y += line_h
    return y


def draw_centered(draw, text, font, cx, y, fill):
    w = draw.textlength(text, font=font)
    draw.text((cx - w // 2, y), text, font=font, fill=fill)


def horizontal_rule(draw, x0, x1, y, width=1, fill=None):
    fill = fill or RULE
    draw.rectangle([x0, y, x1, y + width - 1], fill=fill)


def crop_center(img, w, h):
    iw, ih = img.size
    scale = max(w / iw, h / ih)
    nw, nh = int(iw * scale), int(ih * scale)
    img = img.resize((nw, nh), Image.LANCZOS)
    left = (nw - w) // 2
    top  = (nh - h) // 2
    return img.crop((left, top, left + w, top + h))


def darken(img, factor=0.45):
    """Escurece a imagem misturando com preto."""
    black = Image.new("RGB", img.size, (0, 0, 0))
    return Image.blend(img, black, factor)


# ── Blocos de conteúdo fictício (para áreas com blur) ─────────────────────

SIDEBAR_ITEMS = [
    ("POLÍTICA",
     "Parlamento debate proposta de lei sobre ética e inteligência artificial",
     "Iniciativa do Governo pretende criar um quadro legal para o uso de sistemas autónomos no setor público e privado.",
     "Página 6"),
    ("ECONOMIA",
     "Empresas portuguesas aceleram investimento em automação responsável",
     "Estudo indica crescimento sustentado, mas alerta para necessidade de formação e governança tecnológica.",
     "Página 10"),
    ("SOCIEDADE",
     "Confiança dos cidadãos na tecnologia depende de transparência e controlo democrático",
     "Sondagem revela que 72% dos portugueses defendem regras claras antes da expansão da robótica em serviços públicos.",
     "Página 12"),
]

STRIP_ITEMS = [
    ("SEGURANÇA",
     "Polícia testa robôs de apoio em operações não letais",
     "Projeto-piloto em três cidades avalia eficácia e impacto na relação com os cidadãos."),
    ("INTERNACIONAL",
     "União Europeia prepara regulamento comum para IA de alto risco",
     "Cimeira em Bruxelas reúne ministros para debater prazos de implementação."),
    ("OPINIÃO",
     "Tecnologia ao serviço do bem comum",
     "Por Helena Duarte — a inovação precisa de uma bússola ética que não pode ser deixada ao mercado."),
    ("AMBIENTE",
     "Robótica e sustentabilidade: inovação pode ajudar a enfrentar a crise climática",
     "Relatório aponta para redução de 18% nas emissões com automação industrial inteligente."),
]


# ── Função principal ────────────────────────────────────────────────────────

def gerar_capa(
    titulo: str,
    lead: str,
    legenda: str,
    imagem_path: str | None = None,
    nome_jornal: str = "Observador Cívico",
    output: str = "capa_jornal.png",
    largura: int = 900,
):
    """
    Gera a capa do jornal.

    Parâmetros
    ----------
    titulo        : manchete principal
    lead          : parágrafo introdutório da notícia
    legenda       : texto da caixa de destaque (pull quote / legenda)
    imagem_path   : caminho para a imagem principal (ou None para placeholder)
    nome_jornal   : nome do jornal (ficará com blur)
    output        : caminho do ficheiro PNG a guardar
    largura       : largura total em píxeis (altura calculada automaticamente)
    """

    # ── Dimensões ──────────────────────────────────────────────────────────
    W          = largura
    MARGIN     = int(W * 0.018)       # margem exterior
    COL_MAIN   = int(W * 0.62)        # largura coluna principal
    COL_SIDE   = W - COL_MAIN - MARGIN * 3   # largura sidebar
    GUTTER     = MARGIN               # espaço entre colunas

    # Alturas estimadas — crescem conforme conteúdo
    MAST_H     = int(W * 0.095)
    SUBLINE_H  = int(W * 0.028)
    IMG_H      = int(W * 0.30)
    STRIP_H    = int(W * 0.13)

    # Fontes escaladas com a largura
    s = W / 900   # factor de escala
    f_masthead   = load_font(SERIF_BOLD,   int(56 * s))
    f_subline    = load_font(SANS_REG,     int(13 * s))
    f_headline   = load_font(SERIF_BOLD,   int(38 * s))
    f_lead       = load_font(SERIF_REG,    int(15 * s))
    f_caption    = load_font(SERIF_ITALIC, int(13 * s))
    f_pullquote  = load_font(SERIF_ITALIC, int(14 * s))
    f_cat        = load_font(SANS_BOLD,    int(11 * s))
    f_side_head  = load_font(SERIF_BOLD,   int(15 * s))
    f_side_body  = load_font(SANS_REG,     int(12 * s))
    f_side_pg    = load_font(SANS_REG,     int(11 * s))
    f_strip_cat  = load_font(SANS_BOLD,    int(10 * s))
    f_strip_head = load_font(SERIF_BOLD,   int(13 * s))
    f_strip_body = load_font(SANS_REG,     int(11 * s))

    # ── Estimativa de altura total ─────────────────────────────────────────
    # Vamos desenhar num canvas alto e cortar no final
    CANVAS_H = int(W * 1.55)
    img_full = Image.new("RGB", (W, CANVAS_H), PAPER)
    draw     = ImageDraw.Draw(img_full)

    y = 0  # cursor vertical

    # ══════════════════════════════════════════════════════════════════════
    # 1. MASTHEAD (ficará com blur depois)
    # ══════════════════════════════════════════════════════════════════════
    mast_region = (0, 0, W, MAST_H + SUBLINE_H + MARGIN)

    draw_centered(draw, nome_jornal.upper(), f_masthead,
                  W // 2, MARGIN // 2, fill=INK)
    y_sub = MAST_H - int(4 * s)
    horizontal_rule(draw, MARGIN, W - MARGIN, y_sub, width=1)
    y_sub += int(3 * s)

    today = datetime.date.today()
    dias   = ["segunda-feira","terça-feira","quarta-feira","quinta-feira",
              "sexta-feira","sábado","domingo"]
    meses  = ["janeiro","fevereiro","março","abril","maio","junho",
              "julho","agosto","setembro","outubro","novembro","dezembro"]
    date_str = f"{dias[today.weekday()]}, {today.day} de {meses[today.month-1]} de {today.year}"

    draw.text((MARGIN, y_sub), date_str, font=f_subline, fill=MUTED)
    mid_txt = "Ano II · Número 324 · 1,70 € (IVA incluído)"
    draw_centered(draw, mid_txt, f_subline, W // 2, y_sub, fill=MUTED)
    draw.text((W - MARGIN - int(draw.textlength("Diretor: M. Santos", font=f_subline)),
               y_sub), "Diretor: M. Santos", font=f_subline, fill=MUTED)

    y_sub += int(18 * s)
    horizontal_rule(draw, MARGIN, W - MARGIN, y_sub, width=1)
    y_sub += int(3 * s)

    # Guardar y de início do corpo
    y_body_start = y_sub

    # ══════════════════════════════════════════════════════════════════════
    # 2. LINHA DIVISÓRIA DO CORPO
    # ══════════════════════════════════════════════════════════════════════
    horizontal_rule(draw, 0, W, y_body_start, width=int(3 * s))
    y = y_body_start + int(3 * s) + MARGIN

    # ══════════════════════════════════════════════════════════════════════
    # 3. COLUNA PRINCIPAL (notícia principal — nítida)
    # ══════════════════════════════════════════════════════════════════════
    x_main = MARGIN
    main_inner_w = COL_MAIN - MARGIN * 2

    # Título/manchete
    headline_lines = wrap_lines(draw, titulo, f_headline, main_inner_w)
    y = draw_text_block(draw, headline_lines, f_headline,
                        x_main, y, INK, line_spacing=1.15)
    y += int(8 * s)

    # Lead
    lead_lines = wrap_lines(draw, lead, f_lead, main_inner_w)
    y = draw_text_block(draw, lead_lines, f_lead, x_main, y, INK, line_spacing=1.6)
    y += int(10 * s)

    # Imagem
    y_img = y
    if imagem_path and os.path.exists(imagem_path):
        photo = Image.open(imagem_path).convert("RGB")
        photo = crop_center(photo, COL_MAIN - MARGIN, IMG_H)
        photo = darken(photo, factor=0.35)
        img_full.paste(photo, (x_main, y_img))
    else:
        # placeholder cinzento
        draw.rectangle([x_main, y_img,
                         x_main + COL_MAIN - MARGIN, y_img + IMG_H],
                        fill=(200, 196, 185))
        ph_txt = "[ imagem ]"
        ph_w   = int(draw.textlength(ph_txt, font=f_lead))
        draw.text((x_main + (COL_MAIN - MARGIN - ph_w) // 2,
                   y_img + IMG_H // 2 - int(10 * s)),
                  ph_txt, font=f_lead, fill=(140, 135, 125))
    y = y_img + IMG_H + int(6 * s)

    # Caixa de legenda/destaque
    leg_lines = wrap_lines(draw, legenda, f_caption, main_inner_w - int(16 * s))
    leg_h     = len(leg_lines) * int(20 * s) + int(16 * s)
    draw.rectangle([x_main, y, x_main + COL_MAIN - MARGIN, y + leg_h],
                   fill=PULLBG)
    horizontal_rule(draw, x_main, x_main + COL_MAIN - MARGIN, y,     width=int(2 * s))
    horizontal_rule(draw, x_main, x_main + COL_MAIN - MARGIN, y + leg_h - int(2 * s), width=int(2 * s))
    draw_text_block(draw, leg_lines, f_caption,
                    x_main + int(8 * s), y + int(8 * s), INK, line_spacing=1.5)
    y = y + leg_h + MARGIN

    y_main_end = y  # onde a coluna principal termina

    # ══════════════════════════════════════════════════════════════════════
    # 4. SIDEBAR (ficará com blur depois)
    # ══════════════════════════════════════════════════════════════════════
    x_side  = x_main + COL_MAIN + GUTTER
    y_side  = y_body_start + int(3 * s) + MARGIN

    # linha vertical de separação
    draw.rectangle([x_side - GUTTER // 2, y_body_start,
                    x_side - GUTTER // 2, CANVAS_H],
                   fill=(180, 172, 155))

    for cat, head, body, pg in SIDEBAR_ITEMS:
        # categoria
        draw.text((x_side, y_side), cat, font=f_cat, fill=MUTED)
        y_side += int(16 * s)
        # título
        head_lines = wrap_lines(draw, head, f_side_head, COL_SIDE)
        y_side = draw_text_block(draw, head_lines, f_side_head,
                                 x_side, y_side, INK, line_spacing=1.2)
        y_side += int(5 * s)
        # corpo
        body_lines = wrap_lines(draw, body, f_side_body, COL_SIDE)
        y_side = draw_text_block(draw, body_lines, f_side_body,
                                 x_side, y_side, MUTED, line_spacing=1.5)
        y_side += int(4 * s)
        # página
        draw.text((x_side, y_side), pg, font=f_side_pg, fill=MUTED)
        y_side += int(14 * s)
        # divisor
        horizontal_rule(draw, x_side, x_side + COL_SIDE, y_side,
                         width=1, fill=(180, 172, 155))
        y_side += int(10 * s)

    # ══════════════════════════════════════════════════════════════════════
    # 5. TIRA INFERIOR (ficará com blur)
    # ══════════════════════════════════════════════════════════════════════
    y_strip = max(y_main_end, y_side) + MARGIN
    horizontal_rule(draw, 0, W, y_strip, width=int(3 * s))
    y_strip += int(3 * s)

    strip_col_w = (W - MARGIN * 2) // 4
    for i, (cat, head, body) in enumerate(STRIP_ITEMS):
        xs = MARGIN + i * strip_col_w
        ys = y_strip + MARGIN // 2
        draw.text((xs, ys), cat, font=f_strip_cat, fill=MUTED)
        ys += int(14 * s)
        hlines = wrap_lines(draw, head, f_strip_head, strip_col_w - MARGIN)
        ys = draw_text_block(draw, hlines, f_strip_head, xs, ys, INK, 1.2)
        ys += int(4 * s)
        blines = wrap_lines(draw, body, f_strip_body, strip_col_w - MARGIN)
        draw_text_block(draw, blines, f_strip_body, xs, ys, MUTED, 1.4)
        if i < 3:
            draw.rectangle([xs + strip_col_w - 1, y_strip,
                             xs + strip_col_w, y_strip + STRIP_H],
                            fill=(180, 172, 155))

    y_total = y_strip + STRIP_H + MARGIN

    # ══════════════════════════════════════════════════════════════════════
    # 6. CORTAR canvas ao conteúdo real
    # ══════════════════════════════════════════════════════════════════════
    img_full = img_full.crop((0, 0, W, y_total))

    # ══════════════════════════════════════════════════════════════════════
    # 7. APLICAR BLUR POR REGIÕES
    # ══════════════════════════════════════════════════════════════════════
    BLUR_R = int(12 * s)

    def blur_region(base_img, x0, y0, x1, y1, radius):
        """Extrai, desenfoca e cola uma região."""
        y1 = min(y1, base_img.height)
        if y1 <= y0:
            return base_img
        region = base_img.crop((x0, y0, x1, y1))
        region = region.filter(ImageFilter.GaussianBlur(radius=radius))
        base_img.paste(region, (x0, y0))
        return base_img

    # Blur: masthead (nome do jornal + subline)
    img_full = blur_region(img_full, 0, 0, W, y_body_start, BLUR_R)

    # Blur: sidebar
    img_full = blur_region(img_full,
                            x_side - GUTTER // 2, y_body_start,
                            W, y_strip,
                            BLUR_R)

    # Blur: tira inferior
    img_full = blur_region(img_full, 0, y_strip, W, img_full.height, BLUR_R)

    # ══════════════════════════════════════════════════════════════════════
    # 8. GUARDAR
    # ══════════════════════════════════════════════════════════════════════
    img_full.save(output, quality=95)
    print(f"✔ Capa guardada em: {output}")
    return output


# ── Exemplo de uso ─────────────────────────────────────────────────────────
if __name__ == "__main__":
    gerar_capa(
        titulo     = "Prevenção de Caos Social",
        lead       = (
            "Especialistas alertam que a introdução desregulada de robôs em áreas críticas da vida pública pode comprometer a paz social. Defendem normas "
            "claras sobre responsabilidade civil e ética antes de uma adoção em larga escala."
        ),
        legenda    = (
            "Juristas e investigadores defendem legislação preventiva para garantir "
            "uma integração pacífica da tecnologia em serviços públicos, segurança "
            "e infraestruturas críticas."
        ),
        imagem_path = None,          # substitui pelo caminho da tua imagem
        nome_jornal = "Observador Cívico",
        output      = "capa_jornal.png",
        largura     = 900,
    )
```

```python
from reachy_mini import ReachyMini
import soundfile as sf
import time
import random
from reachy_mini.utils import create_head_pose

# Dicionário para guardar as instâncias
robos = {}

def conectar_robos():
    # Inicializa sem o 'with' para a conexão não fechar
    r = ReachyMini()
    # Inicia o sistema de áudio apenas UMA vez
    r.media.start_playing()
    return r

# Conecta uma vez e guarda
robos["Francisca"] = conectar_robos()

def animar_fala(robot, duracao):
    """Move a cabeça sutilmente enquanto fala (Versão Síncrona)."""
    # Define o tempo de término
    fim = time.time() + duracao
    
    while time.time() < fim:
        # Pequenos movimentos aleatórios
        r = random.uniform(-20, 20)  # Roll
        print("Passou aqui " + str(r))
        
        
        robot.goto_target(
            head=create_head_pose(z=0, roll=r, degrees=True),
            duration=2
        )
        
    
    
def tocar_audio_imediato(robo, caminho_ficheiro):
    data, samplerate = sf.read(caminho_ficheiro)
    # Envia diretamente para o buffer (sem abrir/fechar o sistema de som)
    robo.media.push_audio_sample(data)
   
    animar_fala(robo, len(data)/48000)
    



# --- Agora podes chamar isto quando quiseres, sem delay ---
tocar_audio_imediato(robos["Francisca"], "test1.wav")
```

```python
import os
import wave
import time
import random
import numpy as np
import soundfile as sf
from reachy_mini import ReachyMini
from reachy_mini.utils import create_head_pose

# --- Parâmetros ---
NETWORK_LEAD   = 0.10   # Latência de rede estimada (segundos) — ajustar com ping
MAX_PITCH      = 12     # Graus máx. para cima/baixo (mais subtil)
MAX_ROLL       = 10     # Graus máx. de inclinação lateral
MAX_YAW        = 14     # Graus máx. de rotação horizontal
RMS_WINDOW     = 0.4    # Janela para calcular volume médio (segundos)

# Duração das poses em função do volume (segundos)
POSE_DUR_QUIET = (1.5, 3.0)   # silêncio ou fala suave
POSE_DUR_MID   = (0.9, 1.6)   # fala normal
POSE_DUR_LOUD  = (0.5, 0.9)   # fala alta ou ênfase


def _rms_envelope(data: np.ndarray, samplerate: int, window: float) -> np.ndarray:
    """RMS por janela — devolve um valor por janela ao longo do áudio."""
    n = int(samplerate * window)
    frames = len(data) // n
    env = np.zeros(frames)
    for i in range(frames):
        chunk = data[i * n : (i + 1) * n]
        env[i] = float(np.sqrt(np.mean(chunk ** 2)))
    return env


def _random_pose(prev: tuple, intensity: float) -> tuple:
    """Gera uma nova pose suave, nunca oposta à anterior.

    Usa um offset em relação à pose anterior em vez de valores completamente
    aleatórios, para garantir continuidade e evitar saltos bruscos.
    """
    scale = 0.3 + 0.7 * min(intensity * 4, 1.0)  # 0.3 em silêncio, 1.0 no máximo

    def step(prev_val, max_angle):
        # Deslocamento máximo proporcional à escala e ao ângulo máximo
        delta = random.uniform(-max_angle * 0.5, max_angle * 0.5) * scale
        new_val = prev_val + delta
        # Mantém dentro dos limites
        return float(np.clip(new_val, -max_angle, max_angle))

    prev_pitch, prev_roll, prev_yaw = prev
    return (
        step(prev_pitch, MAX_PITCH),
        step(prev_roll,  MAX_ROLL),
        step(prev_yaw,   MAX_YAW),
    )


def _build_keyframes(data: np.ndarray, samplerate: int, wav_duration: float):
    """Constrói a lista de keyframes (timestamp, pose) antes de tocar.

    A duração de cada pose depende do volume nesse momento:
    - silêncio: pose dura mais (robô quase parado)
    - fala alta: pose dura menos (robô mais expressivo)
    """
    env = _rms_envelope(data, samplerate, RMS_WINDOW)
    # Normaliza o envelope entre 0 e 1
    env_max = env.max() if env.max() > 0 else 1.0
    env_norm = env / env_max

    keyframes = []  # lista de (t_início, pitch, roll, yaw, duração)
    t = 0.0
    prev_pose = (0.0, 0.0, 0.0)  # começa centrado

    while t < wav_duration:
        # Volume aproximado neste momento
        env_idx = min(int(t / RMS_WINDOW), len(env_norm) - 1)
        intensity = float(env_norm[env_idx])

        # Duração desta pose em função do volume
        if intensity < 0.25:
            dur = random.uniform(*POSE_DUR_QUIET)
        elif intensity < 0.65:
            dur = random.uniform(*POSE_DUR_MID)
        else:
            dur = random.uniform(*POSE_DUR_LOUD)

        # Não ultrapassar o fim do áudio
        dur = min(dur, wav_duration - t)

        pose = _random_pose(prev_pose, intensity)
        keyframes.append((t, pose, dur))
        prev_pose = pose
        t += dur

    return keyframes


def play_audio_keyframe(mini, wav_path: str):
    wav_path = os.path.abspath(wav_path)

    # 1. Carrega e analisa antes de tocar
    data, samplerate = sf.read(wav_path)
    if data.ndim > 1:
        data = data.mean(axis=1)

    with wave.open(wav_path, "rb") as wf:
        wav_duration = wf.getnframes() / mini.media.get_output_audio_samplerate()

    keyframes = _build_keyframes(data, samplerate, wav_duration)
    print(f"A tocar '{os.path.basename(wav_path)}' ({wav_duration:.1f}s, {len(keyframes)} poses)")

    # 2. Começa a tocar o áudio
    mini.media.play_sound(wav_path)
    t0 = time.monotonic()

    for (kf_t, pose, dur) in keyframes:
        pitch, roll, yaw = pose

        # Envia o goto_target com NETWORK_LEAD de antecedência
        send_at = t0 + kf_t - NETWORK_LEAD
        wait = send_at - time.monotonic()
        if wait > 0:
            time.sleep(wait)

        # Verifica se o áudio já acabou
        if time.monotonic() - t0 >= wav_duration:
            break

        mini.goto_target(
            head=create_head_pose(
                pitch=pitch,
                roll=roll,
                yaw=yaw,
                degrees=True,
            ),
            duration=dur,  # O robô interpola durante exatamente esta duração
        )

    # Reset suave
    mini.cancel_move()
    mini.goto_target(head=create_head_pose(), duration=1.5)
    print("Fim.")
    time.sleep(1.5)


# --- Uso ---
with ReachyMini(host="reachy-mini-one.local") as mini:
    play_audio_keyframe(mini, "test1.wav")
```

```python
import time
import numpy as np
import soundfile as sf
from reachy_mini import ReachyMini
from reachy_mini.utils import create_head_pose

AUDIO_FILE = "test1.wav"

STEP_TIME = 0.2   # 🔥 MUITO importante (5 Hz)
SENSITIVITY = 1.0


def play_audio_with_head_motion(robot, file_path):
    data, samplerate = sf.read(file_path)

    if len(data.shape) > 1:
        data = data.mean(axis=1)

    robot.media.start_playing()

    samples_per_step = int(samplerate * STEP_TIME)

    for i in range(0, len(data), samples_per_step):
        start_time = time.monotonic()

        chunk = data[i:i + samples_per_step]

        # 🎧 enviar áudio
        robot.media.push_audio_sample(chunk)

        if len(chunk) > 0:
            rms = np.sqrt(np.mean(chunk**2))

            pitch = (rms ** 0.5) * SENSITIVITY * 50
            print(pitch)
            roll = np.random.uniform(-50, 50) * rms
            yaw = np.random.uniform(-50, 50) * rms

            # ✅ movimento LENTO (sem spam!)
            robot.goto_target(
                head=create_head_pose(
                    roll=roll,
                    pitch=pitch,
                    yaw=yaw,
                    degrees=True
                ),
                duration=STEP_TIME  # igual ao intervalo!
            )

        # ⏱️ sync
        elapsed = time.monotonic() - start_time
        sleep_time = STEP_TIME - elapsed

        if sleep_time > 0:
            time.sleep(sleep_time)

    # reset
    robot.goto_target(head=create_head_pose(), duration=1.0)


if __name__ == "__main__":
    with ReachyMini(host="reachy-mini-one.local") as mini:
        print("Ligado")

        mini.goto_target(head=create_head_pose(), duration=1.0)

        play_audio_with_head_motion(mini, AUDIO_FILE)

        print("Done")
```

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# =========================
# CONFIG
# =========================



# =========================
# CENTER CROP (IMPORTANTE)
# =========================
def crop_center(img, target_w, target_h):
    img_ratio = img.width / img.height
    target_ratio = target_w / target_h

    # cortar largura
    if img_ratio > target_ratio:
        new_width = int(img.height * target_ratio)
        offset = (img.width - new_width) // 2
        img = img.crop((offset, 0, offset + new_width, img.height))
    else:
        new_height = int(img.width / target_ratio)
        offset = (img.height - new_height) // 2
        img = img.crop((0, offset, img.width, offset + new_height))

    return img.resize((target_w, target_h))


# =========================
# WRAP TEXTO
# =========================
def wrap_text(text, font, max_width, draw):
    lines = []
    words = text.split()

    while words:
        line = ""
        while words:
            test_line = line + words[0] + " "
            bbox = draw.textbbox((0, 0), test_line, font=font)
            w = bbox[2] - bbox[0]

            if w <= max_width:
                line = test_line
                words.pop(0)
            else:
                break

        lines.append(line.strip())

    return lines


def criar_card(titulo, quote, imagem_path, output):

    WIDTH, HEIGHT = 540, 540
    
    img = Image.open("./articles/background/economic.jpg").convert("RGB")
    img = crop_center(img, WIDTH, HEIGHT)

    # escurecer ligeiramente (melhor leitura)
    black = Image.new("RGB", (WIDTH, HEIGHT), (0, 0, 0))
    img = Image.blend(img, black, 0.8)

    draw = ImageDraw.Draw(img)

    titulo_font = ImageFont.truetype("./Lato/Lato-Bold.ttf", 30)
    quote_font = ImageFont.truetype("./Lato/Lato-Regular.ttf", 20)

    max_width = 500

    titulo_lines = wrap_text(titulo, titulo_font, max_width, draw)

    y = 50

    for line in titulo_lines:
        bbox = draw.textbbox((0, 0), line, font=titulo_font)
        text_width = bbox[2] - bbox[0]

        x = (WIDTH - text_width) / 2

        draw.text((x, y), line, font=titulo_font, fill="white")
        y += 50

    quote_text = f"“{quote}”"
    quote_lines = wrap_text(quote_text, quote_font, max_width, draw)

    line_height = 25
    total_height = len(quote_lines) * line_height

    y_quote = HEIGHT - total_height - 90

    for line in quote_lines:
        bbox = draw.textbbox((0, 0), line, font=quote_font)
        text_width = bbox[2] - bbox[0]

        x = (WIDTH - text_width) / 2

        draw.text((x, y_quote), line, font=quote_font, fill="white")
        y_quote += line_height

    img.save(output)
    print(f"✔ Guardado: {output}")

# Dados para gerar 4 imagens
cards = [
    ("Robôs como Facilitadores de Ligação", 'Discordo totalmente, em vez de isolarem, os robôs podem atuar como facilitadores, ajudando pessoas que já estão isoladas a ligarem-se melhor ao mundo e a outros humanos. A tecnologia pode servir de ponte para quem tem dificuldades sociais, funcionando como um treino para interações reais. O robô não substitui o humano, mas prepara o terreno para novas socializações.', "#1a1a1a"),
    ("Breaking News", "Algo importante aconteceu", "#333366"),
    ("Atualidade", "Frase impactante aqui", "#660000"),
    ("Tecnologia", "O futuro chegou", "#003366"),
]

# Gerar imagens
for i, (titulo, quote, cor) in enumerate(cards):
    criar_card(titulo, quote, cor, f"./articles/card_{i+1}.png")
```

```python

import json
import threading
import webbrowser
from http.server import BaseHTTPRequestHandler, HTTPServer

_server = None
_state = {}
_lock = threading.Lock()


HTML = """<!DOCTYPE html>
<html lang="pt">
<head>
<meta charset="UTF-8">
<title>Mesa Redonda</title>
<link href="https://fonts.googleapis.com/css2?family=IBM+Plex+Mono:wght@400;600&family=IBM+Plex+Sans:wght@300;400;500&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{box-sizing:border-box;margin:0;padding:0}
body{
  font-family:'IBM Plex Sans',sans-serif;
  background:#222;
  color:#f0ede8;
  min-height:100vh;
  display:flex;
  flex-direction:column;
  align-items:center;
  justify-content:center;
  padding:32px 16px;
}
.statement{
  font-size:15px;
  font-weight:300;
  color:#d0d0d0;
  text-align:center;
  max-width:600px;
  margin-bottom:48px;
  line-height:1.6;
  letter-spacing:0.01em;
}
.statement strong{
  display:block;
  font-size:11px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.15em;
  text-transform:uppercase;
  color:#555;
  margin-bottom:8px;
}

/* Mesa redonda */
.table-wrap{
  position:relative;
  width:420px;
  height:420px;
}
.table-circle{
  position:absolute;
  top:50%; left:50%;
  transform:translate(-50%,-50%);
  width:160px; height:160px;
  border-radius:50%;
  border:1px solid #2a2a2a;
  background:#141414;
  display:flex;
  align-items:center;
  justify-content:center;
}
.table-label{
  font-size:10px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.12em;
  text-transform:uppercase;
  color:#333;
}

/* Robôs */
.robot{
  position:absolute;
  width:110px;
  text-align:center;
  transform:translate(-50%,-50%);
  transition:all .4s ease;
}
.robot-icon{
  width:64px; height:64px;
  border-radius:50%;
  border:2px solid #2a2a2a;
  background:#1a1a1a;
  margin:0 auto 10px;
  display:flex;
  align-items:center;
  justify-content:center;
  font-size:26px;
  transition:all .5s ease;
  position:relative;
}
.robot-icon img{
  width:100%; height:100%;
  border-radius:50%;
  object-fit:cover;
}
.robot-icon::after{
  content:'';
  position:absolute;
  inset:-4px;
  border-radius:50%;
  border:2px solid transparent;
  transition:border-color .5s ease;
}
.robot.agree .robot-icon{
  background:#0d2a1a;
  border-color:#1a5c34;
}
.robot.agree .robot-icon img{
  width:100%; height:100%;
  border-radius:50%;
  object-fit:cover;
}
.robot-icon::after{
  border-color:#1d7a45;
}
.robot.disagree .robot-icon{
  background:#2a0d0d;
  border-color:#5c1a1a;
}
.robot.disagree .robot-icon img{
  width:100%; height:100%;
  border-radius:50%;
  object-fit:cover;
}
.robot-icon::after{
  border-color:#7a1d1d;
}

.robot-name{
  font-size:12px;
  font-family:'IBM Plex Mono',monospace;
  color:#d0d0d0;
  margin-bottom:6px;
  transition:color .4s;
}
.robot.agree .robot-name{ color:#3d9e66; }
.robot.disagree .robot-name{ color:#9e3d3d; }

.robot-badge{
  display:inline-block;
  font-size:10px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.05em;
  padding:3px 10px;
  border-radius:20px;
  border:1px solid #222;
  color:#d0d0d0;
  background:#111;
  transition:all .5s ease;
  white-space:nowrap;
}
.robot.agree .robot-badge{
  background:#0d2a1a;
  border-color:#1d7a45;
  color:#3dbc78;
}
.robot.disagree .robot-badge{
  background:#2a0d0d;
  border-color:#7a1d1d;
  color:#bc3d3d;
}

#robot-posicao1{ top:62%; left:80%; }
#robot-posicao2{ top:29%; left:70%; }
#robot-posicao3{ top:29%; left:29.8%; }
#robot-posicao4{ top:62%; left:19.5%; }

/* indicador do participante (estático, no fundo) */
.participant{
  position:absolute;
  top:90.5%; left:50%;
  transform:translate(-50%,-50%);
  text-align:center;
}
.participant-icon{
  width:52px; height:52px;
  border-radius:50%;
  border:2px dashed #2a2a2a;
  background:#111;
  margin:0 auto 6px;
  display:flex;
  align-items:center;
  justify-content:center;
  font-size:20px;
}
.participant-label{
  font-size:10px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.1em;
  text-transform:uppercase;
  color:#333;
}

/* Legenda */
.legend{
  display:flex;
  gap:24px;
  margin-top:40px;
}
.legend-item{
  display:flex;
  align-items:center;
  gap:8px;
  font-size:11px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.08em;
  color:#444;
}
.legend-dot{
  width:10px; height:10px;
  border-radius:50%;
}
.legend-dot.agree{ background:#1d7a45; }
.legend-dot.disagree{ background:#7a1d1d; }
</style>
</head>
<body>

<div class="statement">
  <strong>Statement em discussão</strong>
  <span id="stmt-text">A carregar...</span>
</div>

<div class="table-wrap" id="table-wrap">
  <div class="table-circle">
    <span class="table-label">Mesa<br>Redonda</span>
  </div>
  <div class="participant">
    <div class="participant-icon">&#128100;</div>
    <div class="participant-label">Participante</div>
  </div>
</div>


<script>

let robots = [];
let rendered = false;

function renderRobots(names){
  const wrap = document.getElementById('table-wrap');
  // manter o círculo central
  const circle = wrap.querySelector('.table-circle');
  wrap.innerHTML = '';
  wrap.appendChild(circle);

  robots = names;
  names.forEach((name, i) => {
    const div = document.createElement('div');
    div.className = 'robot';
    div.id = 'robot-' + name;
    div.innerHTML = `
      <div class="robot-icon"><img src="./images/${name}.jpg" alt="${name}"></div>
      <div class="robot-name">${name}</div>
      <div class="robot-badge">a pensar...</div>
    `;
    wrap.appendChild(div);
  });
  rendered = true;
}

function updateRobot(name, side, label){
  const el = document.getElementById('robot-' + name);
  if(!el) return;
  el.className = 'robot ' + side;
  el.querySelector('.robot-badge').textContent = label;
}

async function poll(){
  try{
    const r = await fetch('/state');
    const data = await r.json();

    document.getElementById('stmt-text').textContent = data.statement || '';

    if(!rendered && data.robots && data.robots.length){
      renderRobots(data.robots);
    }

    if(data.responses){
      for(const [name, info] of Object.entries(data.responses)){
        updateRobot(name, info.side, info.label);
      }
    }
  } catch(e){}
  setTimeout(poll, 800);
}

poll();
</script>
</body>
</html>"""


def _make_handler():
    class Handler(BaseHTTPRequestHandler):
        def log_message(self, fmt, *args): pass

        def do_GET(self):
            if self.path == "/" or self.path == "/index.html":
                data = HTML.encode("utf-8")
                self.send_response(200)
                self.send_header("Content-Type", "text/html; charset=utf-8")
                self.end_headers()
                self.wfile.write(data)
        
            elif self.path == "/state":
                with _lock:
                    payload = json.dumps(_state).encode("utf-8")
                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.end_headers()
                self.wfile.write(payload)
        
            elif self.path.startswith("/images/"):
                # servir imagens da pasta robots/
                file_path = self.path.lstrip("/")  # "robots/Robot1.png"
                try:
                    with open(file_path, "rb") as f:
                        data = f.read()
                    self.send_response(200)
                    self.send_header("Content-Type", "image/png")
                    self.end_headers()
                    self.wfile.write(data)
                except FileNotFoundError:
                    self.send_response(404)
                    self.end_headers()
        
            else:
                self.send_response(404)
                self.end_headers()

        def do_POST(self):
            if self.path == "/update":
                length = int(self.headers.get("Content-Length", 0))
                body = json.loads(self.rfile.read(length))
                robot = body.get("robot", "")
                answer = body.get("answer", "").strip().lower()

                info = SCALE.get(answer, {
                    "label": answer.capitalize(),
                    "side":  "neutral",
                    "value": 0
                })

                with _lock:
                    if "responses" not in _state:
                        _state["responses"] = {}
                    _state["responses"][robot] = info

                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.end_headers()
                self.wfile.write(b'{"ok":true}')

            else:
                self.send_response(404)
                self.end_headers()

    return Handler


def start_roundtable(robot_names: list, statement: str, port: int = 8006):
    """Arranca o servidor e abre o browser."""
    global _server, _state
    _state = {
        "statement": statement,
        "robots": robot_names,
        "responses": {}
    }
    _server = HTTPServer(("0.0.0.0", port), _make_handler())
    t = threading.Thread(target=_server.serve_forever, daemon=True)
    t.start()
    webbrowser.open(f"http://127.0.0.1:{port}")
    print(f"Mesa redonda em http://127.0.0.1:{port}")


def update_robot(robot_name: str, answer: str):
    """Chamar quando um robô responde. answer: qualquer string."""
    with _lock:
        if "responses" not in _state:
            _state["responses"] = {}
        _state["responses"][robot_name] = {"label": answer}
    print(f"  -> {robot_name}: {answer}")



def stop_roundtable():
    """Para o servidor."""
    global _server
    if _server:
        _server.shutdown()
        _server = None
        print("Mesa redonda encerrada.")
```

```python
import os
import numpy as np
import librosa
import soundfile as sf

# -----------------------------
# CONFIGURAÇÕES
# -----------------------------

INPUT_FOLDER = "./audios/reachy/Robot2"
OUTPUT_FOLDER = "./audios/reachy/Robot2"

PITCH_STEPS = +2     # grave/agudo
ROBOT_FREQ = 20      # intensidade robotica
DISTORTION = 2       # distorção

# criar pasta de saída
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# extensões aceitas
audio_extensions = (".wav", ".mp3", ".ogg", ".flac")

# -----------------------------
# PROCESSAR TODOS OS ÁUDIOS
# -----------------------------

for filename in os.listdir(INPUT_FOLDER):

    if not filename.lower().endswith(audio_extensions):
        continue

    input_path = os.path.join(INPUT_FOLDER, filename)

    print(f"Processando: {filename}")

    try:
        # carregar áudio
        y, sr = librosa.load(input_path, sr=None)

        # -----------------------------
        # ALTERAR PITCH
        # -----------------------------
        pitched = librosa.effects.pitch_shift(
            y,
            sr=sr,
            n_steps=PITCH_STEPS
        )

        # -----------------------------
        # EFEITO ROBÓTICO
        # -----------------------------
        t = np.arange(len(pitched)) / sr

        modulator = np.sin(
            2 * np.pi * ROBOT_FREQ * t
        )

        robot = pitched * modulator

        # -----------------------------
        # DISTORÇÃO
        # -----------------------------
        robot = np.tanh(robot * DISTORTION)

        # normalizar
        robot = robot / np.max(np.abs(robot))

        # nome do novo arquivo
        output_filename = f"{filename}_modified.wav"

        output_path = os.path.join(
            OUTPUT_FOLDER,
            output_filename
        )

        sf.write(output_path, robot, sr)

        print(f"Salvo: {output_filename}")

    except Exception as e:
        print(f"Erro em {filename}: {e}")

print("Todos os áudios foram processados!")
```

```python
import os
import numpy as np
import librosa
import soundfile as sf

# -----------------------------
# CONFIG
# -----------------------------

INPUT_FOLDER = "./audios/dash/Robot8"
OUTPUT_FOLDER = "./audios/dash/Robot8"

PITCH_STEPS = 7      # menos agressivo
ROBOT_FREQ = 5       # mais suave
MIX = 0.35            # mistura efeito/original

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

audio_extensions = (".wav", ".mp3", ".ogg", ".flac")

# -----------------------------
# PROCESSAMENTO
# -----------------------------

for filename in os.listdir(INPUT_FOLDER):

    if not filename.lower().endswith(audio_extensions):
        continue

    input_path = os.path.join(INPUT_FOLDER, filename)

    print(f"Processando: {filename}")

    try:

        y, sr = librosa.load(input_path, sr=None)

        # -----------------------------
        # PITCH
        # -----------------------------

        pitched = librosa.effects.pitch_shift(
            y,
            sr=sr,
            n_steps=PITCH_STEPS
        )

        # -----------------------------
        # MODULAÇÃO ROBÓTICA SUAVE
        # -----------------------------

        t = np.arange(len(pitched)) / sr

        modulator = np.sin(
            2 * np.pi * ROBOT_FREQ * t
        )

        robot = pitched * modulator

        # -----------------------------
        # MISTURA COM ORIGINAL
        # -----------------------------

        final_audio = (
            pitched * (1 - MIX)
            + robot * MIX
        )

        # normalizar
        final_audio = final_audio / np.max(np.abs(final_audio))

        output_filename = (
            os.path.splitext(filename)[0]
            + "_modified.wav"
        )

        output_path = os.path.join(
            OUTPUT_FOLDER,
            output_filename
        )

        sf.write(output_path, final_audio, sr)

        print(f"Salvo: {output_filename}")

    except Exception as e:
        print(f"Erro em {filename}: {e}")

print("Concluído!")
```

```python
import os
import numpy as np
import librosa
import soundfile as sf

# -----------------------------
# CONFIG
# -----------------------------
INPUT_FOLDER = "./audios/reachy/Robot1"
OUTPUT_FOLDER = "./audios/reachy/Robot1"
PITCH_STEPS = 2   # semitons
SPEED = 0.85      # < 1.0 = mais lento, > 1.0 = mais rápido

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
audio_extensions = (".wav", ".mp3", ".ogg", ".flac")

# -----------------------------
# PROCESSAMENTO
# -----------------------------
for filename in os.listdir(INPUT_FOLDER):
    if not filename.lower().endswith(audio_extensions):
        continue

    input_path = os.path.join(INPUT_FOLDER, filename)
    print(f"Processando: {filename}")

    try:
        y, sr = librosa.load(input_path, sr=None)

        # Pitch shift
        pitched = librosa.effects.pitch_shift(y, sr=sr, n_steps=PITCH_STEPS)

        # Abrandar sem alterar o pitch
        slowed = librosa.effects.time_stretch(pitched, rate=SPEED)

        # Normalizar
        slowed = slowed / np.max(np.abs(slowed))

        output_filename = os.path.splitext(filename)[0] + "_pitched.wav"
        output_path = os.path.join(OUTPUT_FOLDER, output_filename)

        sf.write(output_path, slowed, sr)
        print(f"Salvo: {output_filename}")

    except Exception as e:
        print(f"Erro em {filename}: {e}")

print("Concluído!")
```

```python
import os
import numpy as np
import librosa
import soundfile as sf

# -----------------------------
# CONFIG
# -----------------------------
INPUT_FOLDER = "./audios/pitch"
OUTPUT_FOLDER = "./audios/pitch"
PITCH_STEPS = 1   # semitons

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
audio_extensions = (".wav", ".mp3", ".ogg", ".flac")

# -----------------------------
# PROCESSAMENTO
# -----------------------------
for filename in os.listdir(INPUT_FOLDER):
    if not filename.lower().endswith(audio_extensions):
        continue

    input_path = os.path.join(INPUT_FOLDER, filename)
    print(f"Processando: {filename}")

    try:
        y, sr = librosa.load(input_path, sr=None)

        pitched = librosa.effects.pitch_shift(
            y,
            sr=sr,
            n_steps=PITCH_STEPS,
            bins_per_octave=24,  # mais resolução = menos artefactos
            res_type='kaiser_best'  # melhor qualidade de resample
        )

        # Normalizar
        pitched = pitched / np.max(np.abs(pitched))

        output_filename = os.path.splitext(filename)[0] + "_pitched.wav"
        output_path = os.path.join(OUTPUT_FOLDER, output_filename)

        sf.write(output_path, pitched, sr)
        print(f"Salvo: {output_filename}")

    except Exception as e:
        print(f"Erro em {filename}: {e}")

print("Concluído!")
```

```python
import json
import os

# ← Mete aqui o caminho para a pasta das imagens
IMAGES_DIR = "./articles/background/"

# ← Mete aqui o caminho para o matrix.json
JSON_PATH = "matrix.json"

with open(JSON_PATH, "r") as f:
    data = json.load(f)

def find_img_paths(node, results):
    if isinstance(node, dict):
        for k, v in node.items():
            if k == "img_path" and isinstance(v, str):
                results.add(v)
            else:
                find_img_paths(v, results)

img_paths = set()
find_img_paths(data, img_paths)

print(f"Total de imagens referenciadas: {len(img_paths)}\n")

missing = []
found = []

for img in sorted(img_paths):
    full_path = os.path.join(IMAGES_DIR, img)
    if os.path.exists(full_path):
        found.append(img)
    else:
        missing.append(img)

print(f"✅ Encontradas: {len(found)}")
print(f"❌ Em falta:    {len(missing)}")

if missing:
    print("\nFicheiros em falta:")
    for img in missing:
        print(f"  - {img}")
else:
    print("\nTodas as imagens existem!")
```
