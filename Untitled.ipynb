```python
import asyncio
from dash.robot import DashRobot, discover_and_connect
import time

async def talking_animation(robot, duracao=2.0):
    """Faz o brilho do olho pulsar durante um tempo determinado."""
    inicio = time.time()
    while time.time() - inicio < duracao:
        # Pulsa de 0 a 255 e volta
        for brilho in [0, 50, 150, 255, 150, 50]:
            if not robot.client.is_connected: break
            await robot.eye_brightness(brilho)
            await asyncio.sleep(0.05)
    # Garante que termina ligado
    await robot.eye_brightness(255)
```

```python
names = ["Fernanda"]
robots = {}

async def connect_robots():
    
    print(f"Finding: {', '.join(nomes)}...")
    
    for nome in names:
        try:
            # Tenta conectar a cada um individualmente para melhor controle
            robo = await discover_and_connect([nome])
            if robo:
                robos[nome] = robo
                print(f"✅ {nome} connected: {robo.address}")
        except Exception as e:
            print(f"❌ error to connect {nome}: {e}")

# Executa a conexão
await connect_robots()
```
