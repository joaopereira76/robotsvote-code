```python
import asyncio
from dash.robot import DashRobot, discover_and_connect
import time


async def talking_animation(robot, duracao=2.0):
    """Faz o brilho do olho pulsar durante um tempo determinado."""
    inicio = time.time()
    while time.time() - inicio < duracao:
        # Pulsa de 0 a 255 e volta
        for brilho in [0, 50, 150, 255, 150, 50]:
            if not robot.client.is_connected: break
            await robot.eye_brightness(brilho)
            await asyncio.sleep(0.05)
    # Garante que termina ligado
    await robot.eye_brightness(255)

async def mode_listening(robot, duration=3.0):
    """Faz o robô entrar em modo de escuta ativa por um tempo definido."""
    end_time = time.time() + duration
    # Inclina a cabeça ligeiramente (sinal biológico de escuta)
    await robot.head_pitch(8) 
    await robot.head_yaw(10)
    await robot.neck_color("cyan")

    while time.time() < end_time:
        # Efeito de 'respiração' com a luz do pescoço (Cyan/Azul claro)
        
        await asyncio.sleep(1)
        
        # Pequeno micro-ajuste aleatório para parecer vivo
        await robot.head_yaw(12)
        await asyncio.sleep(1)
        await robot.head_yaw(0)

    await robot.head_pitch(0) 
    await robot.head_yaw(0)

    
```

```python
#total_opposition_robots = ["Maria", "Joao", "Manuel", "Joana"]
total_opposition_robots = ["Fernanda"]
partial_opposition_robots = ["Fernando", "Ana", "Francisca", "Alberto"]
robots = {}

async def connect_robots():
    
    print(f"Finding: {', '.join(total_opposition_robots)}...")
    
    for nome in total_opposition_robots:
        try:
            # Tenta conectar a cada um individualmente para melhor controle
            robo = await discover_and_connect([nome])
            if robo:
                robots[nome] = robo
                print(f"✅ {nome} connected: {robo.address}")
        except Exception as e:
            print(f"❌ error to connect {nome}: {e}")

# Executa a conexão
await connect_robots()
```

```python
async def monitorar_voz_continua(robot, threshold=40):
    """
    Mantém o robô em escuta constante. Quando detecta som acima do limite,
    ativa a animação de escuta.
    """
    print(f"👂 Modo de escuta contínua ativado para {robot.address}")
    
    try:
        while True:
            # 1. Verifica o nível do microfone (método da classe Robot)
            # O sensor_state é atualizado automaticamente pelo RobotSensors
            nivel_som = robot.get_mic_level()
            print(nivel_som)
            
            if nivel_som > threshold:
                print(f"🎤 Som detectado! Nível: {nivel_som}")
                
                # 2. Chama a sua função de animação de escuta
                # Usamos await para ele terminar a animação antes de voltar a ouvir
                await mode_listening(robot, duration=10.0)
                
                print("👂 Voltando a ouvir...")
            
            # Pequena pausa para não sobrecarregar o processador
            await asyncio.sleep(0.2)
            
    except asyncio.CancelledError:
        print("Escuta finalizada.")
    except Exception as e:
        print(f"Erro na escuta: {e}")

# --- COMO EXECUTAR ---

if "Fernanda" in robots:
    # Criamos uma tarefa que roda em "background"
    # Assim você pode continuar a usar outras células do Notebook
    escuta_task = asyncio.create_task(monitorar_voz_continua(robots["Fernanda"]))
```
