```python
from reachy_mini import ReachyMini
import soundfile as sf
import time
import numpy as np
import argparse
import os
import wave



def play_wav(mini: "ReachyMini", wav_path: str) -> None:
    """Play a wav file using the media play_sound API."""
    wav_path = os.path.abspath(wav_path)
    print(f"Playing {wav_path}...")
    mini.media.play_sound(wav_path)

    with wave.open(wav_path, "rb") as wf:
        wav_duration = wf.getnframes() / mini.media.get_output_audio_samplerate()
    time.sleep(wav_duration)
    print("Playback finished.")

with ReachyMini(host="reachy-mini-one.local") as mini:
    # 1. Carregar o ficheiro (precisa da biblioteca soundfile instalada)
    # O Reachy Mini trabalha melhor com ficheiros .wav
    data, samplerate = sf.read('test1.wav') 

    mini.media.start_playing()

    #if len(data.shape) > 1:
        # Fazemos a média dos dois canais para não perder informação
        #data = np.mean(data, axis=1)

    # 3. Garantir o tipo de dados correto (float32 é o padrão para o buffer)
    #data = data.astype(np.float32)

    # 3. Enviar o áudio para o robô 48000Hz
    play_wav(mini,'test1.wav')

    # 4. Esperar o áudio terminar antes de fechar o código
    duracao = len(data) / samplerate
    time.sleep(duracao)

    # 5. Parar o sistema de som
    mini.media.stop_playing()
```

```python
from reachy_mini import ReachyMini
import soundfile as sf
import time
import random
from reachy_mini.utils import create_head_pose

# Dicionário para guardar as instâncias
robos = {}

def conectar_robos():
    # Inicializa sem o 'with' para a conexão não fechar
    r = ReachyMini()
    # Inicia o sistema de áudio apenas UMA vez
    r.media.start_playing()
    return r

# Conecta uma vez e guarda
robos["Francisca"] = conectar_robos()

def animar_fala(robot, duracao):
    """Move a cabeça sutilmente enquanto fala (Versão Síncrona)."""
    # Define o tempo de término
    fim = time.time() + duracao
    
    while time.time() < fim:
        # Pequenos movimentos aleatórios
        r = random.uniform(-20, 20)  # Roll
        print("Passou aqui " + str(r))
        
        
        robot.goto_target(
            head=create_head_pose(z=0, roll=r, degrees=True),
            duration=2
        )
        
    
    
def tocar_audio_imediato(robo, caminho_ficheiro):
    data, samplerate = sf.read(caminho_ficheiro)
    # Envia diretamente para o buffer (sem abrir/fechar o sistema de som)
    robo.media.push_audio_sample(data)
   
    animar_fala(robo, len(data)/48000)
    



# --- Agora podes chamar isto quando quiseres, sem delay ---
tocar_audio_imediato(robos["Francisca"], "test1.wav")
```

```python
import os
import wave
import time
import numpy as np
import soundfile as sf
from reachy_mini import ReachyMini
from reachy_mini.utils import create_head_pose


# --- Parâmetros para Movimentos Calmos ---
STEP_TIME = 0.2    # Aumentado para 200ms (movimentos menos frenéticos)
SENSITIVITY = 1  # Reduzido (reage menos violentamente ao volume)
MAX_ANGLE = 30     # Limita o ângulo máximo para evitar movimentos bruscos
# -----------------------------------------

def play_audio_calm_motion(mini, wav_path):
    wav_path = os.path.abspath(wav_path)
    
    # 1. Análise local do áudio
    data, samplerate = sf.read(wav_path)
    if len(data.shape) > 1:
        data = data.mean(axis=1)

    with wave.open(wav_path, "rb") as wf:
        wav_duration = wf.getnframes() / mini.media.get_output_audio_samplerate()

    # 2. Play Áudio
    print(f"A tocar {wav_path} com movimentos calmos...")
    mini.media.play_sound(wav_path)
    
    start_playback_time = time.monotonic()
    samples_per_step = int(samplerate * STEP_TIME)
    
    # 3. Loop de Movimento Suave
    for i in range(0, len(data), samples_per_step):
        loop_start = time.monotonic()
        
        if (loop_start - start_playback_time) >= wav_duration:
            break

        chunk = data[i:i + samples_per_step]
        if len(chunk) > 0:
            # RMS suavizado (raiz quadrada para não dar saltos bruscos)
            rms = np.sqrt(np.mean(chunk**2))
            
            # Cálculo de pose mais contido
            # Usamos SENSITIVITY menor e multiplicadores baixos
            pitch_val = (rms ** 0.5) * SENSITIVITY * 25 
            
            # Roll e Yaw agora com limites bem curtos (MAX_ANGLE)
            roll_val = np.random.uniform(-MAX_ANGLE, MAX_ANGLE) * rms
            yaw_val = np.random.uniform(-MAX_ANGLE, MAX_ANGLE) * rms

            mini.goto_target(
                head=create_head_pose(
                    roll=roll_val,
                    pitch=pitch_val,
                    yaw=yaw_val,
                    degrees=True
                ),
                duration=STEP_TIME
            )

        elapsed = time.monotonic() - loop_start
        sleep_time = STEP_TIME - elapsed
        if sleep_time > 0:
            time.sleep(sleep_time)

    mini.cancel_move()

    # Reset suave
    mini.goto_target(head=create_head_pose(), duration=1.5)
    print("Fim.")

# Exemplo de uso:
with ReachyMini() as mini:
    play_audio_calm_motion(mini, 'test1.wav')
```

```python
import time
import numpy as np
import soundfile as sf
from reachy_mini import ReachyMini
from reachy_mini.utils import create_head_pose

AUDIO_FILE = "test1.wav"

STEP_TIME = 0.2   # 🔥 MUITO importante (5 Hz)
SENSITIVITY = 1.0


def play_audio_with_head_motion(robot, file_path):
    data, samplerate = sf.read(file_path)

    if len(data.shape) > 1:
        data = data.mean(axis=1)

    robot.media.start_playing()

    samples_per_step = int(samplerate * STEP_TIME)

    for i in range(0, len(data), samples_per_step):
        start_time = time.monotonic()

        chunk = data[i:i + samples_per_step]

        # 🎧 enviar áudio
        robot.media.push_audio_sample(chunk)

        if len(chunk) > 0:
            rms = np.sqrt(np.mean(chunk**2))

            pitch = (rms ** 0.5) * SENSITIVITY * 50
            print(pitch)
            roll = np.random.uniform(-50, 50) * rms
            yaw = np.random.uniform(-50, 50) * rms

            # ✅ movimento LENTO (sem spam!)
            robot.goto_target(
                head=create_head_pose(
                    roll=roll,
                    pitch=pitch,
                    yaw=yaw,
                    degrees=True
                ),
                duration=STEP_TIME  # igual ao intervalo!
            )

        # ⏱️ sync
        elapsed = time.monotonic() - start_time
        sleep_time = STEP_TIME - elapsed

        if sleep_time > 0:
            time.sleep(sleep_time)

    # reset
    robot.goto_target(head=create_head_pose(), duration=1.0)


if __name__ == "__main__":
    with ReachyMini(host="reachy-mini-one.local") as mini:
        print("Ligado")

        mini.goto_target(head=create_head_pose(), duration=1.0)

        play_audio_with_head_motion(mini, AUDIO_FILE)

        print("Done")
```

```python
import random
import time
import asyncio


async def idle_animation(mini, duration):
    end = time.time() + duration

    while time.time() < end:
        pitch = random.uniform(-8, 8)
        yaw = random.uniform(-8, 8)
        roll = random.uniform(-8, 8)

        duration_anim = random.uniform(1.5, 3.0)

        mini.goto_target(
            head=create_head_pose(pitch=pitch, roll=roll, yaw=yaw, degrees=True),
            duration=duration_anim
        )

        await asyncio.sleep(duration_anim + random.uniform(0.5, 1.5))


async def listening_animation(mini, duration):
    end = time.time() + duration

    while time.time() < end:
        mini.goto_target(
            antennas=[np.deg2rad(-30), np.deg2rad(-15)],
            duration=1.0,
        )

        await asyncio.sleep(1.5)

        mini.goto_target(
            antennas=[np.deg2rad(15), np.deg2rad(30)],
            duration=1.0,
        )

        await asyncio.sleep(1.5)

    mini.goto_target(
            antennas=[np.deg2rad(0), np.deg2rad(0)],
            duration=1.0,
        )


from reachy_mini import ReachyMini
from reachy_mini.utils import create_head_pose

with ReachyMini(host="reachy-mini-one.local") as mini:
    # Look up and tilt head
    task1 = asyncio.create_task(idle_animation(mini,20))
    #task2 = asyncio.create_task(listening_animation(mini,10))
    while (not task1.done()):
        await asyncio.sleep(1)
    
    
```
