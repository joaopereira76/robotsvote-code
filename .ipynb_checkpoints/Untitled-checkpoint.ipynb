## Getting Started...

```python
import asyncio
from dash.robot import DashRobot, discover_and_connect
import time

async def talking_animation(robot, duracao=2.0):
    """Faz o brilho do olho pulsar durante um tempo determinado."""
    inicio = time.time()
    while time.time() - inicio < duracao:
        # Pulsa de 0 a 255 e volta
        for brilho in [0, 50, 150, 255, 150, 50]:
            if not robot.client.is_connected: break
            await robot.eye_brightness(brilho)
            await asyncio.sleep(0.05)
    # Garante que termina ligado
    await robot.eye_brightness(255)

async def mode_listening(robot, duration=3.0):
    """Faz o robô entrar em modo de escuta ativa por um tempo definido."""
    end_time = time.time() + duration
    # Inclina a cabeça ligeiramente (sinal biológico de escuta)
    await robot.head_pitch(8) 
    await robot.head_yaw(10)

    while time.time() < end_time:
        # Efeito de 'respiração' com a luz do pescoço (Cyan/Azul claro)
        await robot.neck_color("cyan")
        await asyncio.sleep(0.8)
        await robot.neck_color("black") # Desliga ou baixa intensidade
        await asyncio.sleep(0.8)
        
        # Pequeno micro-ajuste aleatório para parecer vivo
        await robot.head_yaw(12)
        await asyncio.sleep(0.1)
        await robot.head_yaw(10)
```

```python
names = ["Fernanda"]
robots = {}

async def connect_robots():
    
    print(f"Finding: {', '.join(nomes)}...")
    
    for nome in names:
        try:
            # Tenta conectar a cada um individualmente para melhor controle
            robo = await discover_and_connect([nome])
            if robo:
                robos[nome] = robo
                print(f"✅ {nome} connected: {robo.address}")
        except Exception as e:
            print(f"❌ error to connect {nome}: {e}")

# Executa a conexão
await connect_robots()
```

```python
import asyncio
from dash.robot import DashRobot, discover_and_connect
import time

# Dicionário global para manter os robôs conectados entre as células
robos = {}

async def conectar_equipe():
    nomes = ["Fernanda"] # Substitua pelos nomes reais
    print(f"Buscando por: {', '.join(nomes)}...")
    
    for nome in nomes:
        try:
            # Tenta conectar a cada um individualmente para melhor controle
            robo = await discover_and_connect([nome])
            if robo:
                robos[nome] = robo
                print(f"✅ {nome} conectado em {robo.address}")
        except Exception as e:
            print(f"❌ Falha ao conectar em {nome}: {e}")

# Executa a conexão
await conectar_equipe()
```

```python
if "Fernanda" in robos:
    # 1. Inicia o som
    await robos["Fernanda"].say("my2") 
    # 2. Roda a animação sem travar o código (create_task)
    task = asyncio.create_task(animar_olho(robos["Fernanda"], duracao=10.0))
    # 3. Espera a animação terminar antes de você ir para a próxima célula
    await task
```

## Finalize the study

```python
async def disconnect_robots():
    for nome, r in robos.items():
        await r.eye_brightness(0) 
        await r.disconnect()
    print("All robots disconnected.")

await disconnect_robots()
```
