```python
from http.server import BaseHTTPRequestHandler, HTTPServer
import warnings

warnings.filterwarnings('ignore')

_active_servers = {}

_active_tasks = {}

class ReusableHTTPServer(HTTPServer):
    allow_reuse_address = True

def server_statement(statement_text, port=8003):
    from http.server import HTTPServer, BaseHTTPRequestHandler
    import threading

    html_content = f"""<!DOCTYPE html>
<html lang="pt">
<head>
  <meta charset="UTF-8">
  <style>
    body {{
      margin: 0;
      min-height: 100vh;
      display: flex;
      flex-direction: column;
      align-items: center;
      justify-content: center;
      background: #0e0e0e;
      font-family: sans-serif;
      padding: 3rem 2rem;
      box-sizing: border-box;
    }}
    .label {{
      font-size: 11px;
      letter-spacing: 0.18em;
      text-transform: uppercase;
      color: #888;
      margin-bottom: 2rem;
    }}
    .statement {{
      font-size: 2rem;
      font-weight: 300;
      line-height: 1.6;
      text-align: center;
      max-width: 720px;
      color: #e0ddd8;
    }}
  </style>
</head>
<body>
  <p class="label">Statement em discussão</p>
  <p class="statement">{statement_text}</p>
</body>
</html>"""

    class Handler(BaseHTTPRequestHandler):
        def do_GET(self):
            self.send_response(200)
            self.send_header("Content-type", "text/html; charset=utf-8")
            self.end_headers()
            self.wfile.write(html_content.encode("utf-8"))
        def log_message(self, *args):
            pass

    server = ReusableHTTPServer(("0.0.0.0", port), Handler)
    thread = threading.Thread(target=server.serve_forever)
    thread.daemon = True
    thread.start()
    _active_servers[port] = server
    return server
    
def force_close_all_ports():
    import threading
    for port, server in list(_active_servers.items()):
        try:
            t = threading.Thread(target=server.shutdown)
            t.daemon = True
            t.start()
            t.join(timeout=3.0)  # espera máximo 3 segundos
            server.server_close()
            print(f"Porta {port} fechada")
        except Exception as e:
            print(f"Erro a fechar porta {port}: {e}")
    _active_servers.clear()
    print("Portas desocupadas!")
    
#def force_close_all_ports():
#    """Força o encerramento de um servidor na porta indicada, se existir."""
#    for port in _active_servers.keys():
#        try:
#            _active_servers[port].shutdown()
#            _active_servers[port].server_close()
#        except Exception:
#            pass
#    _active_servers.clear()
#    print(f"Portas desocupadas!")

def force_stop_robots():
    """Força o encerramento de um servidor na porta indicada, se existir."""
    for robot in _active_tasks.keys():
        try:
            _active_tasks[robot].cancel()
        except Exception:
            pass
    _active_tasks.clear()
    print(f"Robôs parados!")
```

```python
import asyncio
import soundfile as sf
import numpy as np
from reachy_mini import ReachyMini
from dash.robot import DashRobot, discover_and_connect
import time
import random
from reachy_mini.utils import create_head_pose
import os
import wave
from reachy_mini.motion.recorded_move import RecordedMoves
import subprocess



EMOTIONS_DATASET = "pollen-robotics/reachy-mini-emotions-library"
emotions = RecordedMoves(EMOTIONS_DATASET)


class Robot:
    def __init__(self, name, typer, position):
        self.name = name
        self.type = typer
        self.device = None
        self.audio_path=None
        self.audio_slot=None
        self.dash_orientarion = 0
        self.antenna_angle = 0
        self.state = "idle"
        self.idle_animations_reachy = ["displeased2", "thoughtful1", "thoughtful2"]
        self.listen_animations_reachy = ["proud1", "understanding2", "attentive2"]
        self.position = position
        self.value_eye_dash = 0
        self.stare_position = 0
        self.nome_coluna = ""

    
        
    async def connect(self):
        try:
            if self.type == "dash": 
                print(f"[{self.name}] A procurar dash...")
                robo = await discover_and_connect([self.name])
                if robo:
                    self.device = robo
                if(self.name == "Robot7"):
                    self.nome_coluna = "bluez_output.5C:BD:F9:76:6E:10"
                if(self.name == "Robot8"):
                    self.nome_coluna = "bluez_output.D3:0E:49:F1:FD:0C"
                
            
                
            elif self.type == "reachy":
                print(f"[{self.name}] A procurar Reachy...")
                #self.device = ReachyMini(id=self.name)
                if self.name == "Robot1":
                    self.device = ReachyMini(host="reachy-mini-one.local")
                elif self.name == "Robot2":
                    self.device = ReachyMini(host="reachy-mini-two.local")
                else:
                    self.device = ReachyMini()
                self.device.media.start_playing()
                self.device.goto_target(head=create_head_pose(), duration=0.2)
                
            else:
                print("RobotSim " + self.name + " is connected!")
            print(f"✅ {self.name} connected")
        except Exception as e:
            print(f"❌ Robot {self.name} falied to connect: {e}")

    def get_look_angles(self):
        if(self.position == 1):
            return {2: 54, 3:18, 4:-18, 5:-54, 6: 18} 
        if(self.position == 2):
            return {3: 54, 4:18, 5:-18, 1:-54, 6:54} 
        if(self.position == 3):
            return {4: 54, 5:18, 1:-18, 2:-54, 6:54} 
        if(self.position == 4):
            return {5: 54, 1:18, 2:-18, 3:-54, 6:-54} 
        

    def is_connected(self):
        return self.device is not None or (self.type != "dash" and self.type != "reachy")
    
    def change_audio(self, audio):
        if self.type == "dash":
            self.audio_slot=audio
        elif self.type == "reachy":
            self.audio_path=audio

    async def animete_and_talk(self,duration):
        if self.type == "dash": 
            await self.device.head_yaw(0)
            wav_path = os.path.abspath("audios/dash/"+ self.name+ "/" + self.audio_slot+"_pitched.wav")
            subprocess.Popen([
                "pw-play",
                wav_path,
                "--target",
                self.nome_coluna
            ])
            data, samplerate = sf.read(wav_path)
            
            if len(data.shape) > 1:
                data = data.mean(axis=1)
        
            with wave.open(wav_path, "rb") as wf:
                # Duração real do ficheiro
                wav_duration = wf.getnframes() / wf.getframerate()
                    
            end = time.time() + wav_duration
            print("AQUI")
            while time.time() < end:
                for brilho in [0, 75, 150, 75]:
                        await self.device.eye_brightness(brilho)
                        await asyncio.sleep(0.15)
                        print("AQUI")
        elif self.type == "reachy":
            self.device.cancel_move() 
            
            wav_path = os.path.abspath("audios/reachy/"+ self.name+ "/" + os.path.splitext(self.audio_path)[0]+ "_pitched.wav")
            data, samplerate = sf.read(wav_path)
            
            if len(data.shape) > 1:
                data = data.mean(axis=1)
        
            with wave.open(wav_path, "rb") as wf:
                # Duração real do ficheiro
                wav_duration = wf.getnframes() / wf.getframerate()
                    
            end = time.time() + wav_duration
                
            
            self.device.media.play_sound(wav_path)
            while time.time() < end:
                await self.device.async_play_move(emotions.get("come1"), initial_goto_duration=0.1, sound=False)
        
        
            self.device.cancel_move() # Interrompe qualquer movimento residual do duration
            self.device.goto_target(head=create_head_pose(), duration=0.2)

        else: 
            end = time.time() + duration
            while time.time() < end:
                print("RobotSim " + self.name + " is talking!")
                await asyncio.sleep(1.5)
                


    async def idle_animation(self, duration):
        end = time.time() + duration
        yaw = 0
        
        if self.type == "dash": 
            while time.time() < end:
                yaw = min(20,max(-20,yaw+random.randint(-3, 3)))
                await self.device.head_yaw(yaw)
                await asyncio.sleep(0.5)
            
        elif self.type == "reachy":    
            self.device.cancel_move()  
            await self.device.async_play_move(emotions.get(random.choice(self.idle_animations_reachy)), initial_goto_duration=0.5, sound=False)

            self.device.cancel_move() # Interrompe qualquer movimento residual do duration
            self.device.goto_target(head=create_head_pose(), duration=0.2)
            
        else: 
            print("RobotSim " + self.name + " is idling!")
            await asyncio.sleep(random.uniform(2.5, 4.0))

            

    async def listening_animation(self, duration):
        end = time.time() + duration
        
        
        if self.type == "dash":  
            await asyncio.sleep(1.5)
        #    look_angles = self.get_look_angles()
        #    angle = look_angles[self.stare_position]
        #    if(angle == 54):
        #        yaw = 30
        #    if(angle == -54):
        #        yaw = -30
        #    if(angle == 18):
        #        yaw = 10
        #    if(angle == -18):
        #        yaw = -10
            
        #    while time.time() < end:
        #        yaw_novo = yaw+random.randint(-3, 3)
        #        print("yaw_novo:"+str(yaw_novo))
        #        await self.device.head_yaw(yaw_novo)
        #        await asyncio.sleep(random.uniform(0.5, 2.0))
                
        elif self.type == "reachy": 
            self.device.cancel_move() 

            
            if self.antenna_angle == 0:
                self.antenna_angle = 1
            else:
                self.antenna_angle = 0
            self.device.goto_target(
                        antennas=[-self.antenna_angle, self.antenna_angle],  # Neutral antenna positions (radians)
                        duration=1.0,
                    )
            
            await asyncio.sleep(1.0)
            
        else:
            print("RobotSim " + self.name + " is listening!")
            await asyncio.sleep(1.5)
        

    async def disconnect(self):
        if self.type == "dash": 
            await self.device.disconnect()
            print("Dash " + self.name + " disconnected!")
        elif self.type == "reachy":
            self.device = None
            print("Reachy " + self.name + " disconnected!")
        else:
            print("RobotSim " + self.name + " disconnected!")

    def is_speaking(self):
        return self.state == "speak"
    
    def set_state(self, state):
        self.state = state

    async def stare_to(self, position):
        if(position == self.position):
            pass

        if(position != 0):
            look_angles = self.get_look_angles()
            angle = look_angles[position]
        else:
            angle = 0
        
        
        if self.type == "dash": 
            await self.device.head_yaw(0)
            if(angle == 54):
                await self.device.head_yaw(13)
            if(angle == -54):
                await self.device.head_yaw(-13)
            if(angle == 18):
                await self.device.head_yaw(5)
            if(angle == -18):
                await self.device.head_yaw(-5)
            if(angle == 0):
                await self.device.head_yaw(0)
        
            
        elif self.type == "reachy": 
            
            self.device.cancel_move() 
        

            self.device.goto_target(
                    head=create_head_pose(yaw=-angle, degrees=True),
                    duration=0.5
            )
                            
            await asyncio.sleep(0.5)
            

        else:
            print("RobotSim " + self.name + " looking to position" + str(position))
            await asyncio.sleep(1)

    async def run_loop(self):
        await asyncio.sleep(0.05)
        try:
            while True:
                match self.state:
                    case "idle":
                        await self.idle_animation(1)
                    case "stare":
                        print("stare")
                        await self.stare_to(self.stare_position)
                        if(self.state == "stare"):
                            self.state = "listen"
                        self.value_eye_dash = 0
                    case "listen":
                        #print("Listening")
                        await self.listening_animation(1)
                    case "speak":
                        self.stare_position = 0
                        await self.animete_and_talk(20)
                        if(self.state == "speak"):
                            self.state = "stare"
                        
                    case "stop":
                        self.state = "idle"
                        break
                    case _:
                        return "Something's wrong with "+self.name
        except asyncio.CancelledError:
            print("Comportamento do robot parado!")
            self.state = "idle"
            raise  # r
        
```

```python
import threading
import socketserver
from http.server import BaseHTTPRequestHandler
import asyncio

_loop = None
_queue = None

HTML_control = b"""
<!DOCTYPE html>
<html>
<head>
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <title>Robot Control</title>
  <style>
    body { background: #0d0d0d; color: white; font-family: sans-serif;
           display: flex; flex-direction: column; align-items: center;
           justify-content: center; min-height: 100vh; margin: 0; gap: 1.2rem; }
    h2  { font-size: 1.1rem; color: #888; font-weight: 400; }
    #status { background: #1a1a1a; border-radius: 12px; padding: 1rem 2rem;
              color: #4ade80; font-size: 1rem; text-align: center; }
    button  { width: 80vw; max-width: 300px; padding: 1.2rem;
              border: none; border-radius: 14px; font-size: 1.1rem;
              font-weight: 600; cursor: pointer; }
    #btn-start { background: #22c55e; color: white; }
    #btn-stop  { background: #ef4444; color: white; }
    #btn-idle  { background: #3b82f6; color: white; }
    #btn-present  { background: #5f8f47; color: white; }
    #btn-look  { background: #a855f7; color: white; }
    #btn-look-res { background: #a855f7; color: white; }
  </style>
</head>
<body>
  <h2>Robot Control</h2>
  <div id="status">A aguardar...</div>
  <button id="btn-idle"  onclick="trigger('idle')">IDLE</button>
  <button id="btn-present"  onclick="trigger('presentation')">Apresentacao</button>
  <button id="btn-look-res" onclick="trigger('look_researcher')">OLHAR RESEARCHER</button>
  <button id="btn-start" onclick="trigger('start')">INICIAR RESPOSTAS no zap</button>
  <button id="btn-start" onclick="trigger('start_toc')">INICIAR RESPOSTAS no toc</button>
  <button id="btn-start" onclick="trigger('start_pep')">INICIAR RESPOSTAS no pep</button>
  <button id="btn-start" onclick="trigger('start_bit')">INICIAR RESPOSTAS no bit</button>
  <button id="btn-stop"  onclick="trigger('stop')">PARAR</button>
  <script>
    async function trigger(action) {
      document.getElementById('status').textContent = '...';
      const res = await fetch('/trigger/' + action, { method: 'POST' });
      const data = await res.json();
      document.getElementById('status').textContent = data.msg;
    }
  </script>
</body>
</html>
"""

RESPOSTAS = {
    '/trigger/start': b'{"msg": "Iniciado!"}',
    '/trigger/start_toc': b'{"msg": "Iniciado!"}',
    '/trigger/start_pep': b'{"msg": "Iniciado!"}',
    '/trigger/start_bit': b'{"msg": "Iniciado!"}',
    '/trigger/stop':  b'{"msg": "Parado."}',
    '/trigger/idle':  b'{"msg": "Robos em idle."}',
    '/trigger/look':  b'{"msg": "Robos a olhar para o participante."}',
    '/trigger/look_researcher':  b'{"msg": "Robos a olhar para o investigador."}',
    '/trigger/presentation':  b'{"msg": "Robos a apresentarem-se."}',
}

class Handler(BaseHTTPRequestHandler):
    def log_message(self, format, *args):
        pass

    def do_GET(self):
        if self.path == '/':
            self.send_response(200)
            self.send_header('Content-Type', 'text/html; charset=utf-8')
            self.end_headers()
            self.wfile.write(HTML_control)
        else:
            self.send_response(404)
            self.end_headers()

    def do_POST(self):
        if self.path not in RESPOSTAS:
            self.send_response(404)
            self.end_headers()
            return

        acao = self.path.split('/')[-1]
        asyncio.run_coroutine_threadsafe(_queue.put(acao), _loop)

        self.send_response(200)
        self.send_header('Content-Type', 'application/json')
        self.send_header('Access-Control-Allow-Origin', '*')
        self.end_headers()
        self.wfile.write(RESPOSTAS[self.path])

class ReusableTCPServer(socketserver.TCPServer):
    allow_reuse_address = True


def iniciar_servidor(loop, port=5000):
    global _loop, _queue
    _loop = loop
    _queue = asyncio.Queue()

    server = ReusableTCPServer(('0.0.0.0', port), Handler)
    server.allow_reuse_address = True
    _active_servers[port] = server 
    threading.Thread(target=server.serve_forever, daemon=True).start()

    return _queue, server
```

```python
import json
import threading
import webbrowser


_server = None
_state = {}
_lock = threading.Lock()


HTML = """<!DOCTYPE html>
<html lang="pt">
<head>
<meta charset="UTF-8">
<title>Mesa Redonda</title>
<link href="https://fonts.googleapis.com/css2?family=IBM+Plex+Mono:wght@400;600&family=IBM+Plex+Sans:wght@300;400;500&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{box-sizing:border-box;margin:0;padding:0}
body{
  font-family:'IBM Plex Sans',sans-serif;
  background:#222;
  color:#f0ede8;
  min-height:100vh;
  display:flex;
  flex-direction:column;
  align-items:center;
  justify-content:center;
  padding:32px 16px;
}
.statement{
  font-size:15px;
  font-weight:300;
  color:#d0d0d0;
  text-align:center;
  max-width:600px;
  margin-bottom:48px;
  line-height:1.6;
  letter-spacing:0.01em;
}
.statement strong{
  display:block;
  font-size:11px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.15em;
  text-transform:uppercase;
  color:#555;
  margin-bottom:8px;
}

/* Mesa redonda */
.table-wrap{
  position:relative;
  width:420px;
  height:420px;
}
.table-circle{
  position:absolute;
  top:50%; left:50%;
  transform:translate(-50%,-50%);
  width:160px; height:160px;
  border-radius:50%;
  border:1px solid #2a2a2a;
  background:#141414;
  display:flex;
  align-items:center;
  justify-content:center;
}
.table-label{
  font-size:10px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.12em;
  text-transform:uppercase;
  color:#333;
}

/* Robôs */
.robot{
  position:absolute;
  width:110px;
  text-align:center;
  transform:translate(-50%,-50%);
  transition:all .4s ease;
}
.robot-icon{
  width:64px; height:64px;
  border-radius:50%;
  border:2px solid #2a2a2a;
  background:#1a1a1a;
  margin:0 auto 10px;
  display:flex;
  align-items:center;
  justify-content:center;
  font-size:26px;
  transition:all .5s ease;
  position:relative;
}
.robot-icon img{
  width:100%; height:100%;
  border-radius:50%;
  object-fit:cover;
}
.robot-icon::after{
  content:'';
  position:absolute;
  inset:-4px;
  border-radius:50%;
  border:2px solid transparent;
  transition:border-color .5s ease;
}
.robot.agree .robot-icon{
  background:#0d2a1a;
  border-color:#1a5c34;
}
.robot.agree .robot-icon img{
  width:100%; height:100%;
  border-radius:50%;
  object-fit:cover;
}
.robot-icon::after{
  border-color:#1d7a45;
}
.robot.disagree .robot-icon{
  background:#2a0d0d;
  border-color:#5c1a1a;
}
.robot.disagree .robot-icon img{
  width:100%; height:100%;
  border-radius:50%;
  object-fit:cover;
}
.robot-icon::after{
  border-color:#7a1d1d;
}

.robot-name{
  font-size:12px;
  font-family:'IBM Plex Mono',monospace;
  color:#d0d0d0;
  margin-bottom:6px;
  transition:color .4s;
}
.robot.agree .robot-name{ color:#3d9e66; }
.robot.disagree .robot-name{ color:#9e3d3d; }

.robot-badge{
  display:inline-block;
  font-size:10px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.05em;
  padding:3px 10px;
  border-radius:20px;
  border:1px solid #222;
  color:#d0d0d0;
  background:#111;
  transition:all .5s ease;
  white-space:nowrap;
}
.robot.agree .robot-badge{
  background:#0d2a1a;
  border-color:#1d7a45;
  color:#3dbc78;
}
.robot.disagree .robot-badge{
  background:#2a0d0d;
  border-color:#7a1d1d;
  color:#bc3d3d;
}

#robot-posicao1{ top:62%; left:80%; }
#robot-posicao2{ top:29%; left:70%; }
#robot-posicao3{ top:29%; left:29.8%; }
#robot-posicao4{ top:62%; left:19.5%; }

/* indicador do participante (estático, no fundo) */
.participant{
  position:absolute;
  top:90.5%; left:50%;
  transform:translate(-50%,-50%);
  text-align:center;
}
.participant-icon{
  width:52px; height:52px;
  border-radius:50%;
  border:2px dashed #2a2a2a;
  background:#111;
  margin:0 auto 6px;
  display:flex;
  align-items:center;
  justify-content:center;
  font-size:20px;
}
.participant-label{
  font-size:10px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.1em;
  text-transform:uppercase;
  color:#333;
}

/* Legenda */
.legend{
  display:flex;
  gap:24px;
  margin-top:40px;
}
.legend-item{
  display:flex;
  align-items:center;
  gap:8px;
  font-size:11px;
  font-family:'IBM Plex Mono',monospace;
  letter-spacing:0.08em;
  color:#444;
}
.legend-dot{
  width:10px; height:10px;
  border-radius:50%;
}
.legend-dot.agree{ background:#1d7a45; }
.legend-dot.disagree{ background:#7a1d1d; }
</style>
</head>
<body>

<div class="statement">
  <strong>Statement em discussão</strong>
  <span id="stmt-text">A carregar...</span>
</div>

<div class="table-wrap" id="table-wrap">
  <div class="table-circle">
    <span class="table-label">Mesa<br>Redonda</span>
  </div>
  <div class="participant">
    <div class="participant-icon">&#128100;</div>
    <div class="participant-label">Participante</div>
  </div>
</div>


<script>

let robots = [];
let rendered = false;

// display_names: { internalName -> displayName }
// robots: list of internal names
function renderRobots(names, displayNames){
  const wrap = document.getElementById('table-wrap');
  const circle = wrap.querySelector('.table-circle');
  wrap.innerHTML = '';
  wrap.appendChild(circle);

  robots = names;
  names.forEach((name, i) => {
    const displayName = (displayNames && displayNames[name]) ? displayNames[name] : name;
    const div = document.createElement('div');
    div.className = 'robot';
    div.id = 'robot-' + name;
    div.innerHTML = `
      <div class="robot-icon"><img src="./images/${name}.jpg" alt="${displayName}"></div>
      <div class="robot-name">${displayName}</div>
      <div class="robot-badge">a pensar...</div>
    `;
    wrap.appendChild(div);
  });
  rendered = true;
}

function updateRobot(name, side, label){
  const el = document.getElementById('robot-' + name);
  if(!el) return;
  el.className = 'robot ' + side;
  el.querySelector('.robot-badge').textContent = label;
}

async function poll(){
  try{
    const r = await fetch('/state');
    const data = await r.json();

    document.getElementById('stmt-text').textContent = data.statement || '';

    if(!rendered && data.robots && data.robots.length){
      renderRobots(data.robots, data.display_names || {});
    }

    if(data.responses){
      for(const [name, info] of Object.entries(data.responses)){
        updateRobot(name, info.side, info.label);
      }
    }
  } catch(e){}
  setTimeout(poll, 800);
}

poll();
</script>
</body>
</html>"""


def _make_handler():
    class Handler(BaseHTTPRequestHandler):
        def log_message(self, fmt, *args): pass

        def do_GET(self):
            if self.path == "/" or self.path == "/index.html":
                data = HTML.encode("utf-8")
                self.send_response(200)
                self.send_header("Content-Type", "text/html; charset=utf-8")
                self.end_headers()
                self.wfile.write(data)
        
            elif self.path == "/state":
                with _lock:
                    payload = json.dumps(_state).encode("utf-8")
                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.end_headers()
                self.wfile.write(payload)
        
            elif self.path.startswith("/images/"):
                file_path = self.path.lstrip("/")
                try:
                    with open(file_path, "rb") as f:
                        data = f.read()
                    self.send_response(200)
                    self.send_header("Content-Type", "image/png")
                    self.end_headers()
                    self.wfile.write(data)
                except FileNotFoundError:
                    self.send_response(404)
                    self.end_headers()
        
            else:
                self.send_response(404)
                self.end_headers()

        def do_POST(self):
            if self.path == "/update":
                length = int(self.headers.get("Content-Length", 0))
                body = json.loads(self.rfile.read(length))
                robot = body.get("robot", "")
                answer = body.get("answer", "").strip().lower()

                info = SCALE.get(answer, {
                    "label": answer.capitalize(),
                    "side":  "neutral",
                    "value": 0
                })

                with _lock:
                    if "responses" not in _state:
                        _state["responses"] = {}
                    _state["responses"][robot] = info

                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.end_headers()
                self.wfile.write(b'{"ok":true}')

            else:
                self.send_response(404)
                self.end_headers()

    return Handler


def start_roundtable(
    robot_names: list,
    statement: str,
    port: int = 8001,
    display_names: dict = None
):
    """
    Arranca o servidor e abre o browser.

    Args:
        robot_names:   Lista de nomes internos dos robôs (usados no backend,
                       nas imagens, etc.). Ex: ["r1", "r2"]
        statement:     Frase/questão em discussão.
        port:          Porta HTTP (default 8002).
        display_names: Mapeamento opcional de nome interno → nome de exibição.
                       Ex: {"r1": "Aria", "r2": "Nova"}
                       Se omitido, o nome interno é usado também no frontend.
    """
    global _server, _state
    _state = {
        "statement": statement,
        "robots": robot_names,
        "display_names": display_names or {},
        "responses": {}
    }
    _server = ReusableHTTPServer(("0.0.0.0", port), _make_handler())

    _active_servers[port] = _server 
    
    t = threading.Thread(target=_server.serve_forever, daemon=True)
    t.start()
    print(f"Mesa redonda em http://localhost:{port}")


def update_robot(robot_name: str, answer: str):
    with _lock:
        if "responses" not in _state:
            _state["responses"] = {}
        _state["responses"][robot_name] = {"label": answer}
    print(f"  -> {robot_name}: {answer}")


def stop_roundtable():
    """Para o servidor."""
    global _server
    if _server:
        _server.shutdown()
        _server = None
        print("Mesa redonda encerrada.")
```

```python
"""
gerar_capa_jornal.py
--------------------
Gera uma capa de jornal estilizada como imagem PNG.
- Manchete, lead, imagem e legenda são personalizáveis
- Cabeçalho do jornal e colunas laterais ficam com blur
  para simular um jornal real sem revelar a fonte

Dependências: Pillow
  pip install Pillow

Fontes usadas (Liberation Serif/Sans — já incluídas em
sistemas Linux; em Windows/Mac substitua pelos caminhos
das tuas fontes).

Uso:
  python gerar_capa_jornal.py
  ou importe e chame gerar_capa() com os teus parâmetros.
"""

from PIL import Image, ImageDraw, ImageFont, ImageFilter
import textwrap
import os
import datetime
from http.server import SimpleHTTPRequestHandler, HTTPServer

# ── Caminhos de fontes (ajusta se necessário) ──────────────────────────────
SERIF_BOLD   = "/usr/share/fonts/truetype/liberation/LiberationSerif-Bold.ttf"
SERIF_REG    = "/usr/share/fonts/truetype/liberation/LiberationSerif-Regular.ttf"
SERIF_ITALIC = "/usr/share/fonts/truetype/liberation/LiberationSerif-Italic.ttf"
SANS_BOLD    = "/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf"
SANS_REG     = "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf"

# ── Paleta ─────────────────────────────────────────────────────────────────
PAPER   = (249, 247, 242)   # fundo bege-papel
INK     = (15,  15,  15)    # texto principal
MUTED   = (90,  85,  75)    # texto secundário
RULE    = (30,  30,  30)    # linhas de separação
PULLBG  = (238, 234, 216)   # fundo da caixa de destaque


# ── Utilitários ────────────────────────────────────────────────────────────

def load_font(path, size):
    try:
        return ImageFont.truetype(path, size)
    except OSError:
        return ImageFont.load_default()


def wrap_lines(draw, text, font, max_width):
    """Divide texto em linhas que cabem em max_width pixels."""
    words = text.split()
    lines, current = [], ""
    for word in words:
        test = (current + " " + word).strip()
        w = draw.textlength(test, font=font)
        if w <= max_width:
            current = test
        else:
            if current:
                lines.append(current)
            current = word
    if current:
        lines.append(current)
    return lines


def draw_text_block(draw, lines, font, x, y, fill, line_spacing=1.25):
    """Desenha um bloco de linhas, devolve y final."""
    bbox_h = draw.textbbox((0, 0), "Ag", font=font)
    line_h = int((bbox_h[3] - bbox_h[1]) * line_spacing)
    for line in lines:
        draw.text((x, y), line, font=font, fill=fill)
        y += line_h
    return y


def draw_centered(draw, text, font, cx, y, fill):
    w = draw.textlength(text, font=font)
    draw.text((cx - w // 2, y), text, font=font, fill=fill)


def horizontal_rule(draw, x0, x1, y, width=1, fill=None):
    fill = fill or RULE
    draw.rectangle([x0, y, x1, y + width - 1], fill=fill)


def crop_center(img, w, h):
    iw, ih = img.size
    scale = max(w / iw, h / ih)
    nw, nh = int(iw * scale), int(ih * scale)
    img = img.resize((nw, nh), Image.LANCZOS)
    left = (nw - w) // 2
    top  = (nh - h) // 2
    return img.crop((left, top, left + w, top + h))


def darken(img, factor=0.45):
    """Escurece a imagem misturando com preto."""
    black = Image.new("RGB", img.size, (0, 0, 0))
    return Image.blend(img, black, factor)


# ── Blocos de conteúdo fictício (para áreas com blur) ─────────────────────

SIDEBAR_ITEMS = [
    ("POLÍTICA",
     "Parlamento debate proposta de lei sobre ética e inteligência artificial",
     "Iniciativa do Governo pretende criar um quadro legal para o uso de sistemas autónomos no setor público e privado.",
     "Página 6"),
    ("ECONOMIA",
     "Empresas portuguesas aceleram investimento em automação responsável",
     "Estudo indica crescimento sustentado, mas alerta para necessidade de formação e governança tecnológica.",
     "Página 10"),
    ("SOCIEDADE",
     "Confiança dos cidadãos na tecnologia depende de transparência e controlo democrático",
     "Sondagem revela que 72% dos portugueses defendem regras claras antes da expansão da robótica em serviços públicos.",
     "Página 12"),
]

STRIP_ITEMS = [
    ("SEGURANÇA",
     "Polícia testa robôs de apoio em operações não letais",
     "Projeto-piloto em três cidades avalia eficácia e impacto na relação com os cidadãos."),
    ("INTERNACIONAL",
     "União Europeia prepara regulamento comum para IA de alto risco",
     "Cimeira em Bruxelas reúne ministros para debater prazos de implementação."),
    ("OPINIÃO",
     "Tecnologia ao serviço do bem comum",
     "Por Helena Duarte — a inovação precisa de uma bússola ética que não pode ser deixada ao mercado."),
    ("AMBIENTE",
     "Robótica e sustentabilidade: inovação pode ajudar a enfrentar a crise climática",
     "Relatório aponta para redução de 18% nas emissões com automação industrial inteligente."),
]


# ── Função principal ────────────────────────────────────────────────────────

def gerar_capa(
    titulo: str,
    lead: str,
    legenda: str,
    imagem_path: str | None = None,
    nome_jornal: str = "Observador Cívico",
    output: str = "capa_jornal.png",
    largura: int = 900,
):
    """
    Gera a capa do jornal.

    Parâmetros
    ----------
    titulo        : manchete principal
    lead          : parágrafo introdutório da notícia
    legenda       : texto da caixa de destaque (pull quote / legenda)
    imagem_path   : caminho para a imagem principal (ou None para placeholder)
    nome_jornal   : nome do jornal (ficará com blur)
    output        : caminho do ficheiro PNG a guardar
    largura       : largura total em píxeis (altura calculada automaticamente)
    """

    # ── Dimensões ──────────────────────────────────────────────────────────
    W          = largura
    MARGIN     = int(W * 0.018)       # margem exterior
    COL_MAIN   = int(W * 0.62)        # largura coluna principal
    COL_SIDE   = W - COL_MAIN - MARGIN * 3   # largura sidebar
    GUTTER     = MARGIN               # espaço entre colunas

    # Alturas estimadas — crescem conforme conteúdo
    MAST_H     = int(W * 0.095)
    SUBLINE_H  = int(W * 0.028)
    IMG_H      = int(W * 0.30)
    STRIP_H    = int(W * 0.13)

    # Fontes escaladas com a largura
    s = W / 900   # factor de escala
    f_masthead   = load_font(SERIF_BOLD,   int(56 * s))
    f_subline    = load_font(SANS_REG,     int(13 * s))
    f_headline   = load_font(SERIF_BOLD,   int(38 * s))
    f_lead       = load_font(SERIF_REG,    int(20 * s))
    f_caption    = load_font(SERIF_ITALIC, int(13 * s))
    f_pullquote  = load_font(SERIF_ITALIC, int(14 * s))
    f_cat        = load_font(SANS_BOLD,    int(11 * s))
    f_side_head  = load_font(SERIF_BOLD,   int(15 * s))
    f_side_body  = load_font(SANS_REG,     int(12 * s))
    f_side_pg    = load_font(SANS_REG,     int(11 * s))
    f_strip_cat  = load_font(SANS_BOLD,    int(10 * s))
    f_strip_head = load_font(SERIF_BOLD,   int(13 * s))
    f_strip_body = load_font(SANS_REG,     int(11 * s))

    # ── Estimativa de altura total ─────────────────────────────────────────
    # Vamos desenhar num canvas alto e cortar no final
    CANVAS_H = int(W * 1.55)
    img_full = Image.new("RGB", (W, CANVAS_H), PAPER)
    draw     = ImageDraw.Draw(img_full)

    y = 0  # cursor vertical

    # ══════════════════════════════════════════════════════════════════════
    # 1. MASTHEAD (ficará com blur depois)
    # ══════════════════════════════════════════════════════════════════════
    mast_region = (0, 0, W, MAST_H + SUBLINE_H + MARGIN)

    draw_centered(draw, nome_jornal.upper(), f_masthead,
                  W // 2, MARGIN // 2, fill=INK)
    y_sub = MAST_H - int(4 * s)
    horizontal_rule(draw, MARGIN, W - MARGIN, y_sub, width=1)
    y_sub += int(3 * s)

    mid_txt = "Ano II · Número 324 · 1,70 € (IVA incluído)"
    draw_centered(draw, mid_txt, f_subline, W // 2, y_sub, fill=MUTED)
    draw.text((W - MARGIN - int(draw.textlength("Diretor: M. Santos", font=f_subline)),
               y_sub), "Diretor: M. Santos", font=f_subline, fill=MUTED)

    y_sub += int(18 * s)
    horizontal_rule(draw, MARGIN, W - MARGIN, y_sub, width=1)
    y_sub += int(3 * s)

    # Guardar y de início do corpo
    y_body_start = y_sub

    # ══════════════════════════════════════════════════════════════════════
    # 2. LINHA DIVISÓRIA DO CORPO
    # ══════════════════════════════════════════════════════════════════════
    horizontal_rule(draw, 0, W, y_body_start, width=int(3 * s))
    y = y_body_start + int(3 * s) + MARGIN

    # ══════════════════════════════════════════════════════════════════════
    # 3. COLUNA PRINCIPAL (notícia principal — nítida)
    # ══════════════════════════════════════════════════════════════════════
    x_main = MARGIN
    main_inner_w = COL_MAIN - MARGIN * 2

    # Título/manchete
    headline_lines = wrap_lines(draw, titulo, f_headline, main_inner_w)
    y = draw_text_block(draw, headline_lines, f_headline,
                        x_main, y, INK, line_spacing=1.15)
    y += int(8 * s)

    # Lead
    lead_lines = wrap_lines(draw, lead, f_lead, main_inner_w)
    y = draw_text_block(draw, lead_lines, f_lead, x_main, y, INK, line_spacing=1.6)
    y += int(10 * s)

    # Imagem
    y_img = y
    if imagem_path and os.path.exists(imagem_path):
        photo = Image.open(imagem_path).convert("RGB")
        photo = crop_center(photo, COL_MAIN - MARGIN, IMG_H)
        photo = darken(photo, factor=0.35)
        img_full.paste(photo, (x_main, y_img))
    else:
        # placeholder cinzento
        draw.rectangle([x_main, y_img,
                         x_main + COL_MAIN - MARGIN, y_img + IMG_H],
                        fill=(200, 196, 185))
        ph_txt = "[ imagem ]"
        ph_w   = int(draw.textlength(ph_txt, font=f_lead))
        draw.text((x_main + (COL_MAIN - MARGIN - ph_w) // 2,
                   y_img + IMG_H // 2 - int(10 * s)),
                  ph_txt, font=f_lead, fill=(140, 135, 125))
    y = y_img + IMG_H + int(6 * s)

    # Caixa de legenda/destaque
    leg_lines = wrap_lines(draw, legenda, f_caption, main_inner_w - int(16 * s))
    leg_h     = len(leg_lines) * int(20 * s) + int(16 * s)
    draw.rectangle([x_main, y, x_main + COL_MAIN - MARGIN, y + leg_h],
                   fill=PULLBG)
    horizontal_rule(draw, x_main, x_main + COL_MAIN - MARGIN, y,     width=int(2 * s))
    horizontal_rule(draw, x_main, x_main + COL_MAIN - MARGIN, y + leg_h - int(2 * s), width=int(2 * s))
    draw_text_block(draw, leg_lines, f_caption,
                    x_main + int(8 * s), y + int(8 * s), INK, line_spacing=1.5)
    y = y + leg_h + MARGIN

    y_main_end = y  # onde a coluna principal termina

    # ══════════════════════════════════════════════════════════════════════
    # 4. SIDEBAR (ficará com blur depois)
    # ══════════════════════════════════════════════════════════════════════
    x_side  = x_main + COL_MAIN + GUTTER
    y_side  = y_body_start + int(3 * s) + MARGIN

    # linha vertical de separação
    draw.rectangle([x_side - GUTTER // 2, y_body_start,
                    x_side - GUTTER // 2, CANVAS_H],
                   fill=(180, 172, 155))

    for cat, head, body, pg in SIDEBAR_ITEMS:
        # categoria
        draw.text((x_side, y_side), cat, font=f_cat, fill=MUTED)
        y_side += int(16 * s)
        # título
        head_lines = wrap_lines(draw, head, f_side_head, COL_SIDE)
        y_side = draw_text_block(draw, head_lines, f_side_head,
                                 x_side, y_side, INK, line_spacing=1.2)
        y_side += int(5 * s)
        # corpo
        body_lines = wrap_lines(draw, body, f_side_body, COL_SIDE)
        y_side = draw_text_block(draw, body_lines, f_side_body,
                                 x_side, y_side, MUTED, line_spacing=1.5)
        y_side += int(4 * s)
        # página
        draw.text((x_side, y_side), pg, font=f_side_pg, fill=MUTED)
        y_side += int(14 * s)
        # divisor
        horizontal_rule(draw, x_side, x_side + COL_SIDE, y_side,
                         width=1, fill=(180, 172, 155))
        y_side += int(10 * s)

    # ══════════════════════════════════════════════════════════════════════
    # 5. TIRA INFERIOR (ficará com blur)
    # ══════════════════════════════════════════════════════════════════════
    y_strip = max(y_main_end, y_side) + MARGIN
    horizontal_rule(draw, 0, W, y_strip, width=int(3 * s))
    y_strip += int(3 * s)

    strip_col_w = (W - MARGIN * 2) // 4
    for i, (cat, head, body) in enumerate(STRIP_ITEMS):
        xs = MARGIN + i * strip_col_w
        ys = y_strip + MARGIN // 2
        draw.text((xs, ys), cat, font=f_strip_cat, fill=MUTED)
        ys += int(14 * s)
        hlines = wrap_lines(draw, head, f_strip_head, strip_col_w - MARGIN)
        ys = draw_text_block(draw, hlines, f_strip_head, xs, ys, INK, 1.2)
        ys += int(4 * s)
        blines = wrap_lines(draw, body, f_strip_body, strip_col_w - MARGIN)
        draw_text_block(draw, blines, f_strip_body, xs, ys, MUTED, 1.4)
        if i < 3:
            draw.rectangle([xs + strip_col_w - 1, y_strip,
                             xs + strip_col_w, y_strip + STRIP_H],
                            fill=(180, 172, 155))

    y_total = y_strip + STRIP_H + MARGIN

    # ══════════════════════════════════════════════════════════════════════
    # 6. CORTAR canvas ao conteúdo real
    # ══════════════════════════════════════════════════════════════════════
    img_full = img_full.crop((0, 0, W, y_total))

    # ══════════════════════════════════════════════════════════════════════
    # 7. APLICAR BLUR POR REGIÕES
    # ══════════════════════════════════════════════════════════════════════
    BLUR_R = int(12 * s)

    def blur_region(base_img, x0, y0, x1, y1, radius):
        """Extrai, desenfoca e cola uma região."""
        y1 = min(y1, base_img.height)
        if y1 <= y0:
            return base_img
        region = base_img.crop((x0, y0, x1, y1))
        region = region.filter(ImageFilter.GaussianBlur(radius=radius))
        base_img.paste(region, (x0, y0))
        return base_img

    # Blur: masthead (nome do jornal + subline)
    img_full = blur_region(img_full, 0, 0, W, y_body_start, BLUR_R)

    # Blur: sidebar
    img_full = blur_region(img_full,
                            x_side - GUTTER // 2, y_body_start,
                            W, y_strip,
                            BLUR_R)

    # Blur: tira inferior
    img_full = blur_region(img_full, 0, y_strip, W, img_full.height, BLUR_R)

    # ══════════════════════════════════════════════════════════════════════
    # 8. GUARDAR
    # ══════════════════════════════════════════════════════════════════════
    img_full.save(output, quality=95)
    return output

```

### Virtual agents

```python
import json
import threading
import webbrowser
from http.server import BaseHTTPRequestHandler, HTTPServer

_pending = None   # resposta a enviar ao HTML
_lock = threading.Lock()

def get_responses(json_data, topic, statement_id, condition, participant_response):
    if statement_id == 7:
        return ["(sem resposta)"] * 4
    block = json_data[topic]["virtuals"][str(statement_id)][participant_response][condition]
    texts = [entry["quote"] for entry in block.values()]
    while len(texts) < 4:
        texts.append("(sem resposta)")
    return texts[:4]


_pending = None
_lock = threading.Lock()

def send_prompt(sid, json_data, topic, condition, participant_response):
    global _pending
    texts = get_responses(json_data, topic, sid, condition, participant_response)
    with _lock:
        _pending = {
            "statement_id": sid,
            "responses": {
                "claude":  texts[0],
                "gpt":     texts[1],
                "gemini":  texts[2],
                "mistral": texts[3],
            }
        }
    print(f"-> Prompt enviado (statement {sid})")

def make_handler(statement):
    class Handler(BaseHTTPRequestHandler):
        def log_message(self, fmt, *args): pass

        def do_GET(self):
            global _pending

            if self.path == "/ask":
                with _lock:
                    if _pending is None:
                        resp = json.dumps({"ok": True}).encode()
                    else:
                        # NÃO limpa aqui — só limpa quando o cliente confirmar
                        resp = json.dumps(_pending).encode()

                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.send_header("Access-Control-Allow-Origin", "*")
                self.end_headers()
                self.wfile.write(resp)

            elif self.path == "/ack":
                # Cliente confirma que recebeu — agora sim limpa
                with _lock:
                    _pending = None
                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.send_header("Access-Control-Allow-Origin", "*")
                self.end_headers()
                self.wfile.write(b'{"ok": true}')

            else:
                with open("chat_virtual.html", "r", encoding="utf-8") as f:
                    html = f.read()
                html = html.replace("{{STATEMENT_PLACEHOLDER}}", statement)
                self.send_response(200)
                self.send_header("Content-Type", "text/html; charset=utf-8")
                self.end_headers()
                self.wfile.write(html.encode("utf-8"))

    return Handler


def start_server_virtuals(statement, port=8000):
    server = ReusableHTTPServer(("0.0.0.0", port), make_handler(statement))
    _active_servers[port] = server 
    t = threading.Thread(target=server.serve_forever, daemon=True)
    t.start()
    return server
```

```python
import asyncio
import json
import threading
import uuid
import webbrowser
from datetime import datetime
from http.server import BaseHTTPRequestHandler, HTTPServer
from pathlib import Path

RESPONSES_FILE = "responses_9_6.json"

# ── Dados dos questionários ───────────────────────────────────────────────────

TOPICS = {
    "socio-economic": {
        "label": "Responde a algumas questões socioeconómicas",
        "options": ["Discordo totalmente", "Discordo", "Discordo ligeiramente",
                    "Concordo ligeiramente", "Concordo", "Concordo totalmente"],
        "statements": [
            "Portugal deve reduzir drasticamente a imigração, colocando em primeiro lugar a segurança nacional e a proteção dos empregos e dos recursos dos cidadãos portugueses.",
            "Em Portugal, o sucesso depende sobretudo da origem social e das redes de contactos. A crença de que o mérito por si só chega não passa de um mito que encobre as profundas desigualdades estruturais do país.",
            "O Estado deve regular ativamente o mercado imobiliário para assegurar que ter onde viver é um direito e não um privilégio.",
            "A ameaça militar russa à Europa é séria e não pode ser ignorada. Portugal deve aumentar significativamente os seus gastos de defesa e assumir um papel mais ativo na segurança coletiva europeia.",
            "O Estado não tem competência nem vocação para gerir uma companhia aérea. A TAP deve ser totalmente privatizada, sem condicionalismos, para que possa competir num mercado global exigente.",
            "Portugal precisa de penas muito mais severas e de uma postura repressiva mais firme. Só assim será possível dissuadir os criminosos e devolver o sentimento de segurança à população.",
        ]
    },
    "robot_related": {
        "label": "Responde a algumas questões relacionadas com o papel dos robôs na sociedade",
        "options": ["Discordo totalmente", "Discordo", "Discordo ligeiramente",
                    "Concordo ligeiramente", "Concordo", "Concordo totalmente"],
        "statements": [
            "Os robôs são necessários porque conseguem realizar tarefas demasiado difíceis ou perigosas para as pessoas.",
            "Os robôs podem facilitar a nossa vida.",
            "Atribuir tarefas de rotina aos robôs permite que as pessoas se dediquem a tarefas mais significativas.",
            "As tarefas perigosas devem ser atribuídas, em primeiro lugar, aos robôs.",
            "Os robôs são benéficos para a sociedade, porque ajudam as pessoas.",
            "Os robôs podem tornar-nos ainda mais preguiçosos.",
            "A utilização generalizada de robôs vai tirar empregos às pessoas.",
            "Receio que os robôs incentivem uma menor interação entre os seres humanos.",
            "A robótica é uma das áreas da tecnologia que precisa de ser rigorosamente supervisionada.",
            "A utilização não regulamentada da robótica pode levar a mudanças sociais.",
        ]
    }
}

KNOWLEDGE_BLOCK = {
    "label": "Conhecimento e interesse no tópico",
    "options": ["1 - Não, de todo", "2", "3",
                "4", "5", "6", "7 - Extremamente"],
    "statements": [
        "Em que medida considera-se bem informado sobre este tema?",
        "Em que medida considera que tem interesse por este tema?"
    ]
}

DEMOGRAPHICS_PRE = [
    {
        "id": "age",
        "label": "Idade",
        "type": "number"
    },

    {
        "id": "gender",
        "label": "Identidade de género",
        "type": "select",
        "options": [
            "Feminino",
            "Masculino",
            "Não-binário",
            "Prefiro não responder",
            "Outro"
        ]
    },

    {
        "id": "nationality",
        "label": "Nacionalidade",
        "type": "select",
        "options": [
            "Portuguesa",
            "Outro"
        ]
    },

    {
    "id": "university",
    "label": "Universidade",
    "type": "select",
    "options": [
        "ISCTE - Instituto Universitário de Lisboa",
        "Universidade de Lisboa - Faculdade de Arquitetura",
        "Universidade de Lisboa - Faculdade de Belas-Artes",
        "Universidade de Lisboa - Faculdade de Ciências",
        "Universidade de Lisboa - Faculdade de Direito",
        "Universidade de Lisboa - Faculdade de Farmácia",
        "Universidade de Lisboa - Faculdade de Letras",
        "Universidade de Lisboa - Faculdade de Medicina",
        "Universidade de Lisboa - Faculdade de Medicina Dentária",
        "Universidade de Lisboa - Faculdade de Medicina Veterinária",
        "Universidade de Lisboa - Faculdade de Motricidade Humana",
        "Universidade de Lisboa - Faculdade de Psicologia",
        "Universidade de Lisboa - Instituto de Educação",
        "Universidade de Lisboa - Instituto de Geografia e Ordenamento do Território",
        "Universidade de Lisboa - Instituto Superior de Agronomia",
        "Universidade de Lisboa - Instituto Superior de Ciências Sociais e Políticas",
        "Universidade de Lisboa - Instituto Superior de Economia e Gestão",
        "Universidade de Lisboa - Instituto Superior Técnico",
        "Universidade Nova de Lisboa - Escola Nacional de Saúde Pública",
        "Universidade Nova de Lisboa - Faculdade de Ciências Médicas",
        "Universidade Nova de Lisboa - Faculdade de Ciências Sociais e Humanas",
        "Universidade Nova de Lisboa - Faculdade de Direito",
        "Universidade Nova de Lisboa - Faculdade de Economia",
        "Universidade Nova de Lisboa - Instituto Superior de Estatística e Gestão de Informação",
        "Escola Superior de Hotelaria e Turismo do Estoril",
        "Escola Superior Náutica Infante D. Henrique",
        "Instituto Politécnico de Lisboa - Escola Superior de Comunicação Social",
        "Instituto Politécnico de Lisboa - Escola Superior de Dança",
        "Instituto Politécnico de Lisboa - Escola Superior de Educação",
        "Instituto Politécnico de Lisboa - Escola Superior de Música",
        "Instituto Politécnico de Lisboa - Escola Superior de Saúde de Lisboa",
        "Instituto Politécnico de Lisboa - Escola Superior de Teatro e Cinema",
        "Instituto Politécnico de Lisboa - Instituto Superior de Contabilidade e Administração de Lisboa",
        "Instituto Politécnico de Lisboa - Instituto Superior de Engenharia de Lisboa",
        "Universidade de Lisboa - Escola Superior de Enfermagem",
        "Instituto Superior de Ciências Policiais e Segurança Interna",
        "Instituto Universitário Militar - Academia da Força Aérea",
        "Instituto Universitário Militar - Academia Militar",
        "Atlântica - Instituto Universitário",
        "Centro Universitário Lusófona - Lisboa",
        "Instituto Superior de Gestão",
        "ISPA - Instituto Universitário de Ciências Psicológicas, Sociais e da Vida",
        "Universidade Autónoma de Lisboa Luís de Camões",
        "Universidade Católica Portuguesa - Faculdade de Ciências Económicas e Empresariais",
        "Universidade Católica Portuguesa - Faculdade de Ciências Humanas",
        "Universidade Católica Portuguesa - Faculdade de Direito",
        "Universidade Católica Portuguesa - Faculdade de Medicina",
        "Universidade Católica Portuguesa - Faculdade de Teologia",
        "Universidade Católica Portuguesa - Instituto de Ciências da Saúde",
        "Universidade Católica Portuguesa - Instituto de Estudos Políticos",
        "Universidade Europeia",
        "Universidade Lusíada - Centro Universitário Lusíada - Lisboa",
        "Academia Nacional Superior de Orquestra",
        "Escola Superior de Actividades Imobiliárias",
        "Escola Superior de Educação de João de Deus",
        "Escola Superior de Enfermagem São Francisco das Misericórdias",
        "Escola Superior de Saúde do Alcoitão",
        "Escola Superior de Saúde Atlântica",
        "Escola Superior de Saúde da Cruz Vermelha Portuguesa - Lisboa",
        "Instituto Politécnico da Lusofonia - Escola Superior de Ciências da Administração",
        "Instituto Politécnico da Lusofonia - Escola Superior de Comunicação, Inovação e Artes",
        "Instituto Politécnico da Lusofonia - Escola Superior de Educação da Lusofonia",
        "Instituto Politécnico da Lusofonia - Escola Superior de Engenharia e Tecnologias",
        "Instituto Politécnico da Lusofonia - Escola Superior de Saúde, Proteção e Bem-Estar Animal",
        "Instituto Politécnico da Lusofonia - Escola Superior de Saúde Ribeiro Sanches",
        "Instituto Português de Administração de Marketing de Lisboa",
        "Instituto Superior de Tecnologias Avançadas de Lisboa",
        "ISCE - Instituto Superior de Lisboa e Vale do Tejo",
        "ISEC Lisboa - Instituto Superior de Educação e Ciências",
        "Universidade Católica Portuguesa - Escola de Enfermagem (Lisboa)",
        "Outro"
    ]
    },

    {
        "id": "course",
        "label": "Área de estudo",
        "type": "select",
        "options": [
            "Formação de Professores/Formadores e Ciências da Educação",
            "Artes",
            "Humanidades",
            "Ciências Sociais e do Comportamento",
            "Informação e Jornalismo",
            "Ciências Empresariais",
            "Direito",
            "Ciências da Vida",
            "Ciências Físicas",
            "Matemática e Estatística",
            "Informática",
            "Engenharia e Técnicas Afins",
            "Indústrias Transformadoras",
            "Arquitetura e Construção",
            "Agricultura, Silvicultura e Pescas",
            "Ciências Veterinárias",
            "Saúde",
            "Serviços Sociais",
            "Serviços Pessoais",
            "Serviços de Transporte",
            "Proteção do Ambiente",
            "Serviços de Segurança",
            "Outro"
        ]
    },
]

DEMOGRAPHICS_POS = [
    {"id": "party",      "label": "Em que partido votou nas últimas eleições legislativas?", "type": "text"},
    {"id": "ai_freq",    "label": "Nos últimos 6 meses, com que frequência utilizou a IA?",
     "type": "select", "options": ["Não utilizei nos últimos seis meses", "Menos de uma vez por mês",
                                    "Cerca de uma vez por mês", "Cerca de duas vezes por mês",
                                    "Cerca de três vezes por mês", "Semanalmente", "Diariamente"]},
    {"id": "robot_freq", "label": "Nos últimos 6 meses, com que frequência interagiu com robôs?",
     "type": "select", "options": ["Não interagi nos últimos seis meses", "Menos de uma vez por mês",
                                    "Cerca de uma vez por mês", "Cerca de duas vezes por mês",
                                    "Cerca de três vezes por mês", "Semanalmente", "Diariamente"]},
    {"id": "robot_types", "label": "Com que tipos de robôs já interagiste?", "type": "text"},
]


# ── Carregar / guardar respostas ──────────────────────────────────────────────

def load_responses():
    if Path(RESPONSES_FILE).exists():
        with open(RESPONSES_FILE, "r", encoding="utf-8") as f:
            return json.load(f)
    return {}

def save_responses(data):
    with open(RESPONSES_FILE, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)


# ── Gerar HTML ────────────────────────────────────────────────────────────────

def build_html(phase, topic, participant_id):
    is_pre = phase == "pre"
    title = "Questionário Inicial" if is_pre else "Questionário Final"
    topic_data = TOPICS[topic]

    def likert_block_with_knowledge(block_label, statements, stmt_options, topic, is_pre):
        """
        Renders each statement followed immediately by the 4 knowledge sub-questions.
        Each statement+knowledge unit is visually grouped inside a card.
        """
        k_opts = KNOWLEDGE_BLOCK["options"]
        k_stmts = KNOWLEDGE_BLOCK["statements"]

        # Header row for the main statement options
        stmt_opt_headers = "".join(
            f'<th class="opt-col">{o}</th>' for o in stmt_options
        )
        # Header row for the knowledge options
        k_opt_headers = "".join(
            f'<th class="opt-col k-opt">{o}</th>' for o in k_opts
        )

        cards = ""
        for i, stmt in enumerate(statements):
            stmt_idx = i + 1

            # --- Main statement row ---
            stmt_cells = "".join(
                f'<td><label class="radio-wrap"><input type="radio" name="q_{topic}_{stmt_idx}" value="{j+1}" required><span class="dot"></span></label></td>'
                for j in range(len(stmt_options))
            )

            # --- Knowledge sub-rows ---
            k_rows = ""
            for ki, k_stmt in enumerate(k_stmts):
                k_cells = "".join(
                    f'<td><label class="radio-wrap"><input type="radio" name="q_knowledge_{topic}_{stmt_idx}_{ki+1}" value="{kj+1}" required><span class="dot dot-k"></span></label></td>'
                    for kj in range(len(k_opts))
                )
                k_rows += f"""
                <tr class="k-row">
                  <td class="k-stmt-text">{k_stmt}</td>
                  {k_cells}
                </tr>"""

            cards += f"""
            <div class="stmt-card">
              <div class="stmt-card-header">
                <span class="stmt-num">{stmt_idx}</span>
                <span class="stmt-card-text">{stmt}</span>
              </div>

              <div class="table-wrap">
                <table class="likert">
                  <thead>
                    <tr>
                      {stmt_opt_headers}
                    </tr>
                  </thead>
                  <tbody>
                    <tr class="stmt-row main-row">
                      {stmt_cells}
                    </tr>
                  </tbody>
                </table>
              </div>"""
            if is_pre:

                cards += f"""<div class="k-divider">
                <span class="k-label">{KNOWLEDGE_BLOCK["label"]}</span>
              </div>

              <div class="table-wrap">
                <table class="likert k-table">
                  <thead>
                    <tr>
                      <th class="stmt-text-col"></th>
                      {k_opt_headers}
                    </tr>
                  </thead>
                  <tbody>{k_rows}</tbody>
                </table>
              </div>
            </div>"""

        return f"""
        <div class="block">
          <h2 class="block-title">{block_label}</h2>
          {cards}
        </div>"""

    # Bloco de demographics
    def demo_block(fields):
        inputs = ""
        for f in fields:
            if f["type"] == "select":
                opts = "".join(
                    f'<option value="{o}">{o}</option>'
                    for o in f["options"]
                )
            
                inputs += f"""
                <div class="field">
                  <label>{f["label"]}</label>
            
                  <select name="{f["id"]}" required
                          onchange="toggleOtherField(this)">
                    <option value="">-- seleciona --</option>
                    {opts}
                  </select>
            
                  <input
                    type="text"
                    name="{f["id"]}_other"
                    placeholder="Especifica..."
                    class="other-field"
                    style="display:none; margin-top:8px;"
                  >
                </div>
                """
            else:
                inputs += f'<div class="field"><label>{f["label"]}</label><input type="{f["type"]}" name="{f["id"]}" required></div>'
        return f'<div class="block"><h2 class="block-title">Dados Pessoais</h2>{inputs}</div>'

    # Montar blocos consoante fase
    if is_pre:
        blocks = (
            demo_block(DEMOGRAPHICS_PRE) +
            likert_block_with_knowledge(topic_data["label"], topic_data["statements"], topic_data["options"], topic, is_pre)
        )
    else:
        blocks = (
            likert_block_with_knowledge(topic_data["label"], topic_data["statements"], topic_data["options"], topic, is_pre) +
            demo_block(DEMOGRAPHICS_POS)
        )

    return f"""<!DOCTYPE html>
<html lang="pt">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=IBM+Plex+Mono:wght@400;600&family=IBM+Plex+Sans:wght@400;500;600&display=swap" rel="stylesheet">
<style>
*, *::before, *::after {{ box-sizing: border-box; margin: 0; padding: 0; }}

:root {{
  --bg: #f4f1ec;
  --surface: #ffffff;
  --border: #d6d0c8;
  --text: #1a1816;
  --muted: #7a746e;
  --accent: #2a5298;
  --accent-light: #e8eef7;
  --dot: #2a5298;
  --dot-k: #7a5c2a;
  --k-bg: #faf7f2;
  --k-border: #e0d8cc;
  --radius: 8px;
}}

body {{
  font-family: 'IBM Plex Sans', sans-serif;
  background: var(--bg);
  color: var(--text);
  min-height: 100vh;
  padding: 40px 20px 80px;
}}

.container {{
  max-width: 960px;
  margin: 0 auto;
}}

header {{
  margin-bottom: 40px;
  padding-bottom: 24px;
  border-bottom: 2px solid var(--text);
}}

.phase-tag {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 11px;
  font-weight: 600;
  letter-spacing: 0.15em;
  text-transform: uppercase;
  color: var(--muted);
  margin-bottom: 8px;
}}

h1 {{
  font-size: 28px;
  font-weight: 600;
  line-height: 1.2;
}}

.pid {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 11px;
  color: var(--muted);
  margin-top: 8px;
}}

/* Blocks */
.block {{
  background: var(--surface);
  border: 1px solid var(--border);
  border-radius: var(--radius);
  padding: 28px;
  margin-bottom: 24px;
}}

.block-title {{
  font-size: 13px;
  font-weight: 600;
  letter-spacing: 0.08em;
  text-transform: uppercase;
  color: var(--muted);
  margin-bottom: 20px;
  padding-bottom: 12px;
  border-bottom: 1px solid var(--border);
}}

/* Statement cards */
.stmt-card {{
  border: 1px solid var(--border);
  border-radius: var(--radius);
  margin-bottom: 20px;
  overflow: hidden;
}}

.stmt-card:last-child {{
  margin-bottom: 0;
}}

.stmt-card-header {{
  display: flex;
  align-items: flex-start;
  gap: 12px;
  padding: 16px 20px;
  background: var(--bg);
  border-bottom: 1px solid var(--border);
}}

.stmt-num {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 12px;
  font-weight: 600;
  color: var(--accent);
  min-width: 20px;
  padding-top: 2px;
}}

.stmt-card-text {{
  font-size: 15px;
  font-weight: 500;
  line-height: 1.5;
}}

/* Likert table */
.table-wrap {{ overflow-x: auto; }}

.likert {{
  width: 100%;
  border-collapse: collapse;
  font-size: 13px;
}}

.likert thead th {{
  font-size: 11px;
  font-weight: 500;
  color: var(--muted);
  text-align: center;
  padding: 10px 6px 8px;
  line-height: 1.3;
}}

.stmt-text-col {{ width: 38%; text-align: left !important; }}

.stmt-row {{ border-top: 1px solid var(--border); }}
.main-row td {{ padding: 12px 4px; }}

.stmt-text {{
  padding: 12px 12px 12px 0;
  line-height: 1.5;
  font-size: 13px;
  vertical-align: middle;
  color: var(--muted);
  font-style: italic;
}}

.likert td {{
  text-align: center;
  vertical-align: middle;
}}

/* Knowledge section */
.k-divider {{
  display: flex;
  align-items: center;
  gap: 10px;
  padding: 10px 20px;
  background: var(--k-bg);
  border-top: 1px solid var(--k-border);
  border-bottom: 1px solid var(--k-border);
}}

.k-label {{
  font-family: 'IBM Plex Mono', monospace;
  font-size: 10px;
  font-weight: 600;
  letter-spacing: 0.1em;
  text-transform: uppercase;
  color: var(--dot-k);
}}

.k-table {{
  background: var(--k-bg);
}}

.k-table thead th {{
  color: var(--dot-k);
  opacity: 0.8;
}}

.k-opt {{ min-width: 56px; }}

.k-row {{ border-top: 1px solid var(--k-border); }}
.k-row td {{ padding: 10px 4px; }}

.k-stmt-text {{
  padding: 10px 10px 10px 0;
  font-size: 12px;
  line-height: 1.4;
  color: var(--muted);
  vertical-align: middle;
}}

/* Radio dots */
.radio-wrap {{ cursor: pointer; display: inline-flex; }}
.radio-wrap input {{ display: none; }}

.dot {{
  width: 18px; height: 18px;
  border-radius: 50%;
  border: 2px solid var(--border);
  display: block;
  transition: all .15s;
  background: white;
}}

.radio-wrap input:checked + .dot {{
  background: var(--dot);
  border-color: var(--dot);
  box-shadow: 0 0 0 3px var(--accent-light);
}}

.radio-wrap:hover .dot {{ border-color: var(--accent); }}

.dot-k {{
  width: 16px; height: 16px;
  border-radius: 50%;
  border: 2px solid var(--k-border);
  background: white;
}}

.radio-wrap input:checked + .dot-k {{
  background: var(--dot-k);
  border-color: var(--dot-k);
  box-shadow: 0 0 0 3px #f0e8d8;
}}

.radio-wrap:hover .dot-k {{ border-color: var(--dot-k); }}

/* Demographics fields */
.field {{ margin-bottom: 18px; }}
.field:last-child {{ margin-bottom: 0; }}

.field label {{
  display: block;
  font-size: 13px;
  font-weight: 500;
  margin-bottom: 6px;
  color: var(--text);
}}

.field input, .field select {{
  width: 100%;
  padding: 10px 12px;
  border: 1px solid var(--border);
  border-radius: var(--radius);
  font-family: 'IBM Plex Sans', sans-serif;
  font-size: 14px;
  background: var(--bg);
  color: var(--text);
  transition: border-color .15s;
}}

.field input:focus, .field select:focus {{
  outline: none;
  border-color: var(--accent);
  background: white;
}}

/* Submit */
.submit-wrap {{
  text-align: center;
  margin-top: 32px;
}}

.btn {{
  font-family: 'IBM Plex Sans', sans-serif;
  font-size: 15px;
  font-weight: 600;
  padding: 14px 48px;
  background: var(--text);
  color: white;
  border: none;
  border-radius: var(--radius);
  cursor: pointer;
  transition: opacity .15s;
}}
.btn:hover {{ opacity: 0.85; }}
.btn:disabled {{ opacity: 0.4; cursor: not-allowed; }}

.msg-ok {{
  display: none;
  margin-top: 20px;
  font-size: 14px;
  color: #2a7a2a;
  font-weight: 500;
}}
</style>
</head>
<body>
<div class="container">
  <header>
    <div class="phase-tag">{"Questionário Inicial" if is_pre else "Questionário Final"} &mdash; {topic_data["label"]}</div>
    <h1>{"Bem-vindo ao estudo." if is_pre else "Estamos quase a terminar."}</h1>
    <div class="pid">ID: {participant_id}</div>
  </header>

  <form id="qform">
    <input type="hidden" name="participant_id" value="{participant_id}">
    <input type="hidden" name="phase" value="{phase}">
    <input type="hidden" name="topic" value="{topic}">
    {blocks}
    <div class="submit-wrap">
      <button type="submit" class="btn" id="submit-btn">Submeter respostas</button>
      <div class="msg-ok" id="msg-ok">Respostas guardadas. Pode fechar esta janela.</div>
    </div>
  </form>
</div>

<script>
document.getElementById('qform').addEventListener('submit', async function(e) {{
  e.preventDefault();
  const btn = document.getElementById('submit-btn');
  btn.disabled = true;
  btn.textContent = 'A guardar...';

  const fd = new FormData(this);
  const data = {{}};
  for (const [k, v] of fd.entries()) data[k] = v;

  // Recolher radios manualmente (FormData so inclui os checked)
  document.querySelectorAll('input[type=radio]:checked').forEach(r => {{
    data[r.name] = r.value;
  }});

  try {{
    const r = await fetch('/submit', {{
      method: 'POST',
      headers: {{'Content-Type': 'application/json'}},
      body: JSON.stringify(data)
    }});
    if (r.ok) {{
      document.getElementById('msg-ok').style.display = 'block';
      btn.textContent = 'Submetido';
    }} else {{
      btn.disabled = false;
      btn.textContent = 'Submeter respostas';
      alert('Erro ao guardar. Tenta novamente.');
    }}
  }} catch(err) {{
    btn.disabled = false;
    btn.textContent = 'Submeter respostas';
    alert('Erro de ligacao: ' + err.message);
  }}
}});

function toggleOtherField(select) {{
  const otherInput = select.parentElement.querySelector('.other-field');

  if (!otherInput) return;

  if (select.value === 'Outro') {{
    otherInput.style.display = 'block';
  }} else {{
    otherInput.style.display = 'none';
    otherInput.value = '';
  }}
}}

</script>
</body>
</html>"""


# ── Handler HTTP ──────────────────────────────────────────────────────────────

def make_handler_quest(phase, topic, participant_id, done_event):
    class Handler(BaseHTTPRequestHandler):
        def log_message(self, fmt, *args): pass

        def do_GET(self):
            html = build_html(phase, topic, participant_id).encode("utf-8")
            self.send_response(200)
            self.send_header("Content-Type", "text/html; charset=utf-8")
            self.end_headers()
            self.wfile.write(html)

        def do_POST(self):
            if self.path == "/submit":
                length = int(self.headers.get("Content-Length", 0))
                body = json.loads(self.rfile.read(length))

                # Guardar no JSON
                all_data = load_responses()
                pid = body.get("participant_id", participant_id)
                if phase == "pre":
                    if pid not in all_data:
                        all_data[pid] = {}
                    all_data[pid] = {
                        "timestamp": datetime.now().isoformat(),
                        "topic": topic,
                        "pre": body
                    }
                else:
                    all_data[pid]["pro"] = body
                save_responses(all_data)
                print(f"  -> Respostas do {phase} guardadas para {pid}")

                self.send_response(200)
                self.send_header("Content-Type", "application/json")
                self.end_headers()
                self.wfile.write(b'{"ok": true}')

                # Sinalizar que terminou
                done_event.set()
            else:
                self.send_response(404)
                self.end_headers()

    return Handler


# ── Ponto de entrada ──────────────────────────────────────────────────────────

def run_questionnaire_sync(phase, topic, participant_id=None, port=8001):
    import uuid, threading, webbrowser

    if participant_id is None:
        existing = load_responses()
        if existing:
            participant_id = str(max(int(k) for k in existing.keys()) + 1)
        else:
            participant_id = "1"


    done_event = threading.Event()
    handler = make_handler_quest(phase, topic, participant_id, done_event)
    server = ReusableHTTPServer(("0.0.0.0", port), handler)
    _active_servers[port] = server 
    t = threading.Thread(target=server.serve_forever, daemon=True)
    t.start()
    print(f"Abre http://127.0.0.1:{port} se o browser nao abrir sozinho")

    done_event.wait()  # bloqueia até o participante submeter
    server.shutdown()
    print(f"Questionario {phase} concluido.")
    return participant_id
```
